## Ejercicio 1: Cargar Datos de ventas.

El archivo datos.dat contiene el registro de las ventas realizadas.

Tiene un formato de ancho fijo:
- fecha:    10 lugares
- producto: 30 lugares
- precio:   10 lugares
- cantidad:  5 lugares

hacer una funcion que carge los en una lista de diccionarios. 

nota: con 
```python
lineas = open("datos.dat", "r").readlines() 
```
puede obtener la lista de string de todos los datos 

````python
def cargar():
    #Obtengo los datos leyendo la informacion del archivo
    lineas = open("datos.dat", "r").readlines()
    #almaceno los datos en la lista
    lista = []
    for item in lineas:

        #Recorro cada uno de los campos de la lista
        Fecha = item[:10].strip()
        Descripcion = item[10:40].strip()
        Stock_Disponible = int(item[40:50].strip())  # Convertir a entero
        Precio_Unitario = float(item[50:55].strip())  # Convertir a flotante
        #Agrego al diccionario los campos
        dicc = {
            "Fecha": Fecha,
            "Descripcion": Descripcion,
            "Stock_Disponible": Stock_Disponible,
            "Precio_Unitario": Precio_Unitario
        }
        #agrego a la lista el diccionario
        lista.append(dicc)
    return lista
listas= cargar();
print(listas);
````

Hacer una funcion que sume los importes vendidos (precio * cantidad) y las cantidades

## Ejercicio 3: Listar las unidades vendidas.

Listar cuantas unidades se vendieron en total para cada producto

````python
def cargar():
    #Obtengo los datos leyendo la informacion del archivo
    lineas = open("datos.dat", "r").readlines()
    #almaceno los datos en la lista
    lista = []
    for item in lineas:

        #Recorro cada uno de los campos de la lista
        Fecha = item[:10].strip()
        Descripcion = item[10:40].strip()
        Stock_Disponible = int(item[40:50].strip())  # Convertir a entero
        Precio_Unitario = float(item[50:55].strip())  # Convertir a flotante
        #Agrego al diccionario los campos
        dicc = {
            "Fecha": Fecha,
            "Descripcion": Descripcion,
            "Stock_Disponible": Stock_Disponible,
            "Precio_Unitario": Precio_Unitario
        }
        #agrego a la lista el diccionario
        lista.append(dicc)
    return lista
listas= cargar();
print(listas);
def unidades_vendidas(lista):
    ventas_por_producto = {}
    #recorre la lista
    for lis in lista:
        #almaceno los datos de producto y cantidad
        producto = lis["Descripcion"]
        cantidad = lis["Stock_Disponible"]
        #recorro el diccionario vacio y acumulo las ventas de cada producto;
        if producto in ventas_por_producto:
            ventas_por_producto[producto] += cantidad  
        else:
            ventas_por_producto[producto] = cantidad  
    return ventas_por_producto
ventas_por_producto = unidades_vendidas(lista);
print(ventas_por_producto)

````

### Ejercicio 5: Ranking de productos

Realizar un listado de los 3 productos mas vendidos ordenados la cantidad de unidades vendidas (ordenadas de mayor a menor)


````python
def cargar():
    #Obtengo los datos leyendo la informacion del archivo
    lineas = open("datos.dat", "r").readlines()
    #almaceno los datos en la lista
    lista = []
    for item in lineas:

        #Recorro cada uno de los campos de la lista
        Fecha = item[:10].strip()
        Descripcion = item[10:40].strip()
        Stock_Disponible = int(item[40:50].strip())  # Convertir a entero
        Precio_Unitario = float(item[50:55].strip())  # Convertir a flotante
        #Agrego al diccionario los campos
        dicc = {
            "Fecha": Fecha,
            "Descripcion": Descripcion,
            "Stock_Disponible": Stock_Disponible,
            "Precio_Unitario": Precio_Unitario
        }
        #agrego a la lista el diccionario
        lista.append(dicc)
    return lista
listas= cargar();
print(listas);
contadores = {}
for venta in listas:
    descripcion = venta["Descripcion"]
    unidades = venta["Stock_Disponible"]
    if descripcion in contadores:
        contadores[descripcion] += unidades
    else:
        contadores[descripcion] = unidades

productos_ordenados = sorted(contadores.items(), key=lambda x: x[1],reverse=True)

# Mostrar los 3 productos más vendidos
top_3 = productos_ordenados[:3]
for producto, unidades in top_3:
    print(f"{producto} - Unidades Vendidas: {unidades}")
````

### Ejercicio 6: Listar las ventas por mes

Realizar un listado del total de unidades vendidas por productos separada por mes.

````python
def cargar():
    #Obtengo los datos leyendo la informacion del archivo
    lineas = open("datos.dat", "r").readlines()
    #almaceno los datos en la lista
    lista = []
    for item in lineas:

        #Recorro cada uno de los campos de la lista
        Fecha = item[:10].strip()
        Descripcion = item[10:40].strip()
        Stock_Disponible = int(item[40:50].strip())  # Convertir a entero
        Precio_Unitario = float(item[50:55].strip())  # Convertir a flotante
        #Agrego al diccionario los campos
        dicc = {
            "Fecha": Fecha,
            "Descripcion": Descripcion,
            "Stock_Disponible": Stock_Disponible,
            "Precio_Unitario": Precio_Unitario
        }
        #agrego a la lista el diccionario
        lista.append(dicc)
    return lista
listas= cargar();
#print(listas);
ventas_por_mes = {
    "agosto": {},
    "septiembre": {}
}

# Realizar un listado del total de unidades vendidas por productos separada por mes
for listi in listas:
    #verifica que la fecha de la list dentro de la lista coincida con un determinado mes
    if "2024-08-01" <= listi["Fecha"] <= "2024-08-31":
        mes = "agosto"
    elif "2024-09-01" <= listi["Fecha"] <= "2024-09-30":
        mes = "septiembre"

    # Acumular las ventas por descripción del producto
    if listi['Descripcion'] in ventas_por_mes[mes]:
        ventas_por_mes[mes][listi['Descripcion']] += listi["Stock_Disponible"]
    else:
        ventas_por_mes[mes][listi['Descripcion']] = listi["Stock_Disponible"]

# Muestro las ventas acumuladas por producto y por mes y luego imprimo 
for mes, productos in ventas_por_mes.items():
    print(f"Ventas en {mes}:\n")
    for producto, total_vendido in productos.items():
        print(f"Producto: {producto} - Unidades vendidas: {total_vendido}\n")
````

### Ejercicio 7: Informe general

Mostrar un listado de productos ordenados alfabeticamente que contengan el precio promedio, la cantidad de unidades vendidas y el importe total vendido para cada producto

````python
def cargar():
    lineas = open("datos.dat", "r").readlines()
    lista = []
    for item in lineas:
        Fecha = item[:10].strip()
        Descripcion = item[10:40].strip()
        Stock_Disponible = int(item[40:50].strip()) 
        Precio_Unitario = float(item[50:55].strip())  

        dicc = {
            "Fecha": Fecha,
            "Descripcion": Descripcion,
            "Stock_Disponible": Stock_Disponible,
            "Precio_Unitario": Precio_Unitario
        }
        lista.append(dicc)
    return lista
lista = cargar();
print(lista)
#Mostrar un listado de productos ordenados alfabeticamente que contengan el precio promedio, 
#la cantidad de unidades vendidas y el importe total vendido para cada producto
productos_info = {}

#recorro la lista y obtengo datos
for venta in lista:
    descripcion = venta["Descripcion"]
    unidades_vendidas = venta["Stock_Disponible"]
    precio_unitario = venta["Precio_Unitario"]

     
    if descripcion not in productos_info:
        productos_info[descripcion] = {
            "total_unidades_vendidas": 0,
            "suma_precio": 0,
            "ventas_totales": 0
        }
    #Acumulo la información para cada producto
    productos_info[descripcion]["total_unidades_vendidas"] += unidades_vendidas
    productos_info[descripcion]["suma_precio"] += precio_unitario
    productos_info[descripcion]["ventas_totales"] += unidades_vendidas * precio_unitario

#ordeno la lista por medio de la descripcion 
productos_ordenados = sorted(productos_info.items())

#Muestro la información por producto
for descripcion, info in productos_ordenados:
    total_unidades = info["total_unidades_vendidas"]
    
    if total_unidades!=0:
        precio_promedio = info["suma_precio"] / total_unidades 
        ventas_totales = info["ventas_totales"]
    
    print(f"Producto: {descripcion}")
    print(f"Unidades vendidas: {total_unidades}")
    print(f"Precio promedio: {precio_promedio:.2f}")
    print(f"Importe total vendido: {ventas_totales:.2f}")
````
