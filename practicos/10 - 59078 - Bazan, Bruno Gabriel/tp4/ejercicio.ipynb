# Simulación del Truco

Las **cartas** deben registra el numero, palo y valor de la carta. 
El valor debe reflejar la jerarquía de las cartas en el juego del truco.

```python
def calcular_valor(numero, palo):
    valores = {
        (1, "espada"): 14,
        (1, "basto"):13,
        (7, "espada"): 12,
        (7, "oro"): 11,
        (3, ): 10,
        (2, ): 9,
        (1, ): 8,
        (12, ): 7,
        (11, ): 6,
        (10, ): 5,
        (7, ): 4,
        (6, ): 3,
        (5, ): 2,
        (4, ): 1
    }
    return valores.get((numero, palo)) or valores.get((numero,))


class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
        self.valor = calcular_valor(numero, palo)

    def __str__(self):
        return f"{self.numero} de {self.palo} vale: ({self.valor}) puntos"




c = Carta(10, "espada")

print(c)
```

**Mazo**: Representan las cartas españolas de 1 al 12, en los palos de oro, copa, espada y basto.

Debe existir una función que permita barajar las cartas.
y otra función que permita repartir las cartas a los jugadores.

```python
import random
numeros = [1, 2, 3, 4, 5, 6, 7, 10, 11, 12]
palos = ["oro", "copa", "espada", "basto"]

class Mazo:
    def __init__(self):
        self.cartas = [Carta(numero, palo) for numero in numeros for palo in palos]

    def mezclar(self):
        random.shuffle(self.cartas)

    def repartir(self, cantidad):
        return [self.cartas.pop() for _ in range(cantidad)]

    def reiniciar(self):
        self.cartas = [Carta(numero, palo) for numero in numeros for palo in palos]
        self.mezclar() 


m = Mazo()
m.mezclar()
cartas_repartidas = m.repartir(3)
```

**Jugador**: Deben tener un nombre y las cartas que tiene en la mano. Debe acumular los puntos ganados.

Nota: Para simplificar el problema, el jugador elige la carta a jugar de forma aleatoria.

```python
class Jugador:

    def __init__(self, nombre):
        self.nombre = nombre

    def recibir_cartas(self, cartas):
        self.cartas = cartas
    
    def jugar_carta(self):
        return self.cartas.pop()
    
    def __str__(self):
        return f"{self.nombre} tiene las cartas: "

j = Jugador("Juan")

j.recibir_cartas(m.repartir(3))

print(j)
for carta in j.cartas:
    print(carta)
```

**Partida**: Debe tener dos jugadores, un mazo de cartas y un puntaje. Debe permitir jugar una carta por ronda y definir el ganador de la ronda. Debe permitir jugar hasta que se terminen las cartas del mazo.

```python
class Partida:
       
    def __init__(self, jugador1, jugador2):
        self.jugador1 = jugador1
        self.jugador2 = jugador2
        self.mazo = Mazo()
        self.mazo.mezclar()
        jugador1.recibir_cartas(self.mazo.repartir(3))
        jugador2.recibir_cartas(self.mazo.repartir(3))
        self.contador_j1 = 0
        self.contador_j2 = 0 

    def jugar_ronda(self):
        carta_j1 = self.jugador1.jugar_carta()
        carta_j2 = self.jugador2.jugar_carta()
        
        print(f"\nEl jugador Juan jugó: {carta_j1}")
        print(f"El jugador Pedro jugó: {carta_j2}")

        if carta_j1.valor > carta_j2.valor:
            self.contador_j1 += 1
            print("Gano el jugador Juan")
        elif carta_j1.valor < carta_j2.valor:
            self.contador_j2 += 1
            print("Gano el jugador Pedro")
        else:
            print("Empate")
        
        # Mostrar puntaje después de cada ronda
        print(f"\nPuntaje actual:")
        print(f"Juan: {self.contador_j1}")
        print(f"Pedro: {self.contador_j2}")

    def jugar(self):
        while self.contador_j1 < 15 and self.contador_j2 < 15:
            self.mazo.reiniciar()
            self.jugador1.recibir_cartas(self.mazo.repartir(3))
            self.jugador2.recibir_cartas(self.mazo.repartir(3))
            self.jugar_ronda()
        
        # Anunciar el ganador final
        if self.contador_j1 >= 15:
            print("\nEl jugador Juan gana la partida!")
        else:
            print("\nEl jugador Pedro gana la partida!")

# Ejemplo de uso
p = Partida(Jugador("Juan"), Jugador("Pedro"))
p.jugar()
```
