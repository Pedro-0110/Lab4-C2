# Simulación del Truco

Las **cartas** deben registra el numero, palo y valor de la carta. 
El valor debe reflejar la jerarquía de las cartas en el juego del truco.

```python

def calcular_valor(numero, palo):
    if numero == 1 and palo == "espada":
        return 14
    elif numero == 1 and palo == "basto":
        return 13
    elif numero == 7 and palo == "espada":
        return 12
    elif numero == 7 and palo == "oro":
        return 11
    elif numero == 3:
        return 10
    elif numero == 2:
        return 9
    elif numero == 1:
        return 8
    elif numero == 12:
        return 7
    elif numero == 11:
        return 6
    elif numero == 10:
        return 5
    elif numero == 7:
        return 4
    elif numero == 6:
        return 3
    elif numero == 5:
        return 2
    elif numero == 4:
        return 1
    else:
        raise ValueError("Número de carta no válido para Truco")

class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
        self.valor = calcular_valor(numero, palo)

    def __repr__(self):
        return f"Carta({self.numero} de {self.palo}, valor={self.valor})"


c = Carta(1, "oro")
print(c)  
```

**Mazo**: Representan las cartas españolas de 1 al 12, en los palos de oro, copa, espada y basto.

Debe existir una función que permita barajar las cartas.
y otra función que permita repartir las cartas a los jugadores.

```python
import random

class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
        self.valor = calcular_valor(numero, palo)

    def __repr__(self):
        return f"{self.numero} de {self.palo}"

def calcular_valor(numero, palo):
    if numero == 1 and palo == "espada":
        return 14
    elif numero == 1 and palo == "basto":
        return 13
    elif numero == 7 and palo == "espada":
        return 12
    elif numero == 7 and palo == "oro":
        return 11
    elif numero == 3:
        return 10
    elif numero == 2:
        return 9
    elif numero == 1:
        return 8
    elif numero == 12:
        return 7
    elif numero == 11:
        return 6
    elif numero == 10:
        return 5
    elif numero == 7:
        return 4
    elif numero == 6:
        return 3
    elif numero == 5:
        return 2
    elif numero == 4:
        return 1
    elif numero in [8, 9]:  # El 8 y 9 no se usan en Truco, podemos asignarles un valor especial o ignorarlos
        return 0
    else:
        raise ValueError("Número de carta no válido para Truco")

class Mazo:
    palos = ["oro", "copa", "espada", "basto"]
    numeros = list(range(1, 13))

    def __init__(self):
        self.cartas = [Carta(numero, palo) for palo in self.palos for numero in self.numeros]

    def mezclar(self):
        random.shuffle(self.cartas)

    def repartir(self, cantidad):
        if cantidad > len(self.cartas):
            raise ValueError("No hay suficientes cartas para repartir.")
        mano = self.cartas[:cantidad]
        self.cartas = self.cartas[cantidad:]
        return mano

m = Mazo()
m.mezclar()
print(m.repartir(3))
```

**Jugador**: Deben tener un nombre y las cartas que tiene en la mano. Debe acumular los puntos ganados.

Nota: Para simplificar el problema, el jugador elige la carta a jugar de forma aleatoria.

```python
import random

class Jugador:
    def __init__(self, nombre):
        self.nombre = nombre
        self.cartas = []
        self.puntos = 0

    def recibir_cartas(self, cartas):
        self.cartas = cartas

    def jugar_carta(self):
        if not self.cartas:
            raise ValueError(f"{self.nombre} no tiene más cartas para jugar.")
        carta_jugada = random.choice(self.cartas)
        self.cartas.remove(carta_jugada)
        return carta_jugada

    def ganar_puntos(self, puntos):
        self.puntos += puntos

    def __repr__(self):
        return f"Jugador {self.nombre} - Puntos: {self.puntos}"

class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
        self.valor = calcular_valor(numero, palo)

    def __repr__(self):
        return f"{self.numero} de {self.palo}"

def calcular_valor(numero, palo):
    if numero == 1 and palo == "espada":
        return 14
    elif numero == 1 and palo == "basto":
        return 13
    elif numero == 7 and palo == "espada":
        return 12
    elif numero == 7 and palo == "oro":
        return 11
    elif numero == 3:
        return 10
    elif numero == 2:
        return 9
    elif numero == 1:
        return 8
    elif numero == 12:
        return 7
    elif numero == 11:
        return 6
    elif numero == 10:
        return 5
    elif numero == 7:
        return 4
    elif numero == 6:
        return 3
    elif numero == 5:
        return 2
    elif numero == 4:
        return 1
    elif numero in [8, 9]:
        return 0
    else:
        raise ValueError("Número de carta no válido para Truco")

class Mazo:
    palos = ["oro", "copa", "espada", "basto"]
    numeros = list(range(1, 13))

    def __init__(self):
        self.cartas = [Carta(numero, palo) for palo in self.palos for numero in self.numeros]

    def mezclar(self):
        random.shuffle(self.cartas)

    def repartir(self, cantidad):
        if cantidad > len(self.cartas):
            raise ValueError("No hay suficientes cartas para repartir.")
        mano = self.cartas[:cantidad]
        self.cartas = self.cartas[cantidad:]
        return mano

m = Mazo()
m.mezclar()

j = Jugador("Juan")
j.recibir_cartas(m.repartir(3))

print(f"Cartas iniciales de {j.nombre}: {j.cartas}")
print(f"{j.nombre} ha jugado: {j.jugar_carta()}")
print(f"Cartas restantes de {j.nombre}: {j.cartas}")

j.ganar_puntos(3)
print(f"{j.nombre} ha ganado puntos. Puntos actuales: {j.puntos}")
```

**Partida**: Debe tener dos jugadores, un mazo de cartas y un puntaje. Debe permitir jugar una carta por ronda y definir el ganador de la ronda. Debe permitir jugar hasta que se terminen las cartas del mazo.
