## Ejercicio 1: Cargar Datos de ventas.

El archivo datos.dat contiene el registro de las ventas realizadas.

Tiene un formato de ancho fijo:
- fecha:    10 lugares
- producto: 30 lugares
- precio:   10 lugares
- cantidad:  5 lugares

hacer una funcion que carge los en una lista de diccionarios. 

nota: con 
```python
lineas = open("datos.dat", "r").readlines() 
```
puede obtener la lista de string de todos los datos 

````python
def cargar_datos():
    lineas = open("datos.dat", "r").readlines()

    ventas = []

    for linea in lineas:
        fecha = linea[0:10].strip()
        producto = linea[10:40].strip()
        precio = float(linea[40:50].strip())
        cantidad = int(linea[50:55].strip())

        venta = {
            'fecha': fecha,
            'producto': producto,
            'precio': precio,
            'cantidad': cantidad
        }

        ventas.append(venta)

    return ventas

datos_ventas = cargar_datos()

for venta in datos_ventas:
    print(f"Fecha: {venta['fecha']}")
    print(f"Producto: {venta['producto']}")
    print(f"Precio: {venta['precio']}")
    print(f"Cantidad: {venta['cantidad']}")
    print('-' * 40) 

````

## Ejercicio 2: Calcular el total de ventas.

Hacer una funcion que sume los importes vendidos (precio * cantidad) y las cantidades

````python
def calcular_totales(datos):
    total_importe = 0  
    total_cantidad = 0  

    for venta in datos:
        importe_venta = venta['precio'] * venta['cantidad'] 
        total_importe += importe_venta  
        total_cantidad += venta['cantidad'] 

    return total_importe, total_cantidad

datos_ventas = cargar_datos()  

importe, cantidad = calcular_totales(datos_ventas)

print(f"Las ventas fueron de ${importe:.2f} en {cantidad} unidades")
````

## Ejercicio 3: Listar las unidades vendidas.

Listar cuantas unidades se vendieron en total para cada producto

````python
def unidades_vendidas(datos):
    ventas_por_producto = {}

    for venta in datos:
        producto = venta['producto']
        
        cantidad = venta['cantidad']

        if producto in ventas_por_producto:
            ventas_por_producto[producto] += cantidad
        else:
            ventas_por_producto[producto] = cantidad

    return ventas_por_producto

def listar_ventas(ventas):
    for producto, cantidad in ventas.items():
        print(f"Producto: {producto}, Unidades Vendidas: {cantidad}")





datos_ventas = cargar_datos() 

ventas = unidades_vendidas(datos_ventas)

listar_ventas(ventas)

````

##  Ejercicio 4: Listar el precio promedio por producto.

Hacer un listado del precio promedio por productos.

````python
def precio_promedio(datos):
    total_precio_por_producto = {}
    cantidad_ventas_por_producto = {}

    for venta in datos:
        producto = venta['producto']
        precio = venta['precio']




        if producto in total_precio_por_producto:
            total_precio_por_producto[producto] += precio
            cantidad_ventas_por_producto[producto] += 1
        else:
            total_precio_por_producto[producto] = precio
            cantidad_ventas_por_producto[producto] = 1

    precio_promedio_por_producto = {}

    for producto in total_precio_por_producto:
        precio_promedio_por_producto[producto] = total_precio_por_producto[producto] / cantidad_ventas_por_producto[producto]

    return precio_promedio_por_producto

def listar_precios(precios):
    for producto, precio in precios.items():
        print(f"Producto: {producto}, Precio Promedio: ${precio:.2f}")

datos_ventas = cargar_datos() 

precios = precio_promedio(datos_ventas)

listar_precios(precios)
````

### Ejercicio 5: Ranking de productos

Realizar un listado de los 3 productos mas vendidos ordenados la cantidad de unidades vendidas (ordenadas de mayor a menor)


````python
def ranking_productos(datos, top=3):
    ventas_por_producto = {}

    for venta in datos:
        producto = venta['producto']
        cantidad = venta['cantidad']

        if producto in ventas_por_producto:
            ventas_por_producto[producto] += cantidad
        else:
            ventas_por_producto[producto] = cantidad

    productos_ordenados = sorted(ventas_por_producto.items(), key=lambda x: x[1], reverse=True)

    return productos_ordenados[:top]

def listar_ranking(ranking):
    for idx, (producto, cantidad) in enumerate(ranking, start=1):
        print(f"{idx}. Producto: {producto}, Unidades Vendidas: {cantidad}")

datos_ventas = cargar_datos()  

ranking = ranking_productos(datos_ventas)

listar_ranking(ranking)
````

### Ejercicio 6: Listar las ventas por mes

Realizar un listado del total de unidades vendidas por productos separada por mes.

````python
from collections import defaultdict
from datetime import datetime

def ventas_por_mes(datos):
    ventas_mensuales = defaultdict(lambda: defaultdict(int))

    for venta in datos:
  
        fecha = datetime.strptime(venta['fecha'], "%Y-%m-%d")
        mes = fecha.strftime("%Y-%m") 

        producto = venta['producto']
        cantidad = venta['cantidad']

        ventas_mensuales[mes][producto] += cantidad

    return ventas_mensuales

def listar_ventas_mensuales(ventas):
 
    for mes, productos in ventas.items():
        print(f"Mes: {mes}")
        for producto, cantidad in productos.items():
            print(f"  Producto: {producto}, Unidades Vendidas: {cantidad}")
        print("-" * 40)  

datos_ventas = cargar_datos()  

ventas_mensuales = ventas_por_mes(datos_ventas)

listar_ventas_mensuales(ventas_mensuales)
````

### Ejercicio 7: Informe general

Mostrar un listado de productos ordenados alfabeticamente que contengan el precio promedio, la cantidad de unidades vendidas y el importe total vendido para cada producto

````python
def resumen_ventas(datos):
    total_precio_por_producto = {}
    total_cantidad_por_producto = {}
    total_importe_por_producto = {}

    for venta in datos:
        producto = venta['producto']
        precio = venta['precio']
        cantidad = venta['cantidad']
        importe = precio * cantidad

        if producto in total_cantidad_por_producto:
            total_precio_por_producto[producto] += precio
            total_cantidad_por_producto[producto] += cantidad
            total_importe_por_producto[producto] += importe
        else:
            total_precio_por_producto[producto] = precio
            total_cantidad_por_producto[producto] = cantidad
            total_importe_por_producto[producto] = importe

    resumen = {}

    for producto in total_precio_por_producto:
        precio_promedio = total_precio_por_producto[producto] / (total_cantidad_por_producto[producto] / cantidad)
        resumen[producto] = {
            'precio_promedio': precio_promedio,
            'cantidad': total_cantidad_por_producto[producto],
            'importe_total': total_importe_por_producto[producto]
        }

    return resumen

def informe_ventas(resumen):
    for producto in sorted(resumen.keys()):
        precio_promedio = resumen[producto]['precio_promedio']
        cantidad = resumen[producto]['cantidad']
        importe_total = resumen[producto]['importe_total']

        print(f"Producto: {producto}")
        print(f"  Precio Promedio: ${precio_promedio:.2f}")
        print(f"  Cantidad Vendida: {cantidad}")
        print(f"  Importe Total: ${importe_total:.2f}")
        print("-" * 40)

datos_ventas = cargar_datos()  
resumen = resumen_ventas(datos_ventas)

informe_ventas(resumen)
````
