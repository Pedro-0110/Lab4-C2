## Ejercicio 1: Cargar Datos de ventas.

El archivo datos.dat contiene el registro de las ventas realizadas.

Tiene un formato de ancho fijo:
- fecha:    10 lugares
- producto: 30 lugares
- precio:   10 lugares
- cantidad:  5 lugares

hacer una funcion que carge los en una lista de diccionarios. 

nota: con 
```python
lineas = open("datos.dat", "r").readlines() 
```
puede obtener la lista de string de todos los datos 

````python
def cargar_datos():

    datos = []
    lineas = open("datos.dat", "r").readlines()

    for indx,value in enumerate(lineas):
       datos.append({
                    "fecha" : value[0:10].strip(),
                    "producto" : value[10:46],
                    "precio" : value[46:53].strip(),
                    "cantidad" : value[53:57].strip()
                    })
    return datos

datos = cargar_datos()
````

## Ejercicio 2: Calcular el total de ventas.

Hacer una funcion que sume los importes vendidos (precio * cantidad) y las cantidades

````python
def calcular_totales(datos):
    importe = 0
    cantidad = 0

    for indx,value in enumerate(datos):
        importe +=  float(datos[indx]["precio"]) * int(datos[indx]["cantidad"])
        cantidad += int(datos[indx]["cantidad"])

    return (importe,cantidad)


importe, cantidad = calcular_totales(datos)
print(f"La ganancia total en ventas fue de ${importe:.2f} en {cantidad} unidades")
````

## Ejercicio 3: Listar las unidades vendidas.

Listar cuantas unidades se vendieron en total para cada producto

````python
def obtener_productos_sin_repetir(datos):
    lista_productos = []
    lista_productos.append(datos[0]["producto"])

    for indx, value in enumerate(datos):
        for indx2, value2 in enumerate(lista_productos):
            if value2 == datos[indx]["producto"]:
                break
            elif value2 != datos[indx]["producto"] and indx2 == len(lista_productos) - 1:
                lista_productos.append(datos[indx]["producto"])

    return lista_productos



def obtener_cantidades_vendidas(productos_sin_repetir):

    lista_cantidades_vendidas = [0] * len(productos_sin_repetir)
    
    for indx,value in enumerate(productos_sin_repetir):
        for indx2,value2 in enumerate(datos):
            if value == datos[indx2]["producto"]:
                    lista_cantidades_vendidas[indx] += int(datos[indx2]["cantidad"])

    return lista_cantidades_vendidas



def unidades_vendidas(datos):
        lista_productos = obtener_productos_sin_repetir(datos)
        lista_unidades_vendidas = obtener_cantidades_vendidas(lista_productos)
        return lista_productos,lista_unidades_vendidas
    


def listar_ventas(datos):
    productos = datos[0]
    ventas = datos[1] 
    print(" UNIDADES VENDIDAS \n")
    print("-----------------------------------------------------------------")
    print("PRODUCTOS                          UNIDADES VENDIDAS")
    print("-----------------------------------------------------------------")
    for indx, value in enumerate(productos):
        
         print( f"{productos[indx]}{str(ventas[indx])}")
         print("_________________________________________________________________")


ventas = unidades_vendidas(datos)
listar_ventas(ventas)
````

##  Ejercicio 4: Listar el precio promedio por producto.

Hacer un listado del precio promedio por productos.

````python

def obtener_ganancia_total_ventas_x_producto(productos_sin_repetir):

    lista_total_venta_producto = []
    lista_total_venta_producto = [0] * len(productos_sin_repetir)

    for indx,value in enumerate(productos_sin_repetir):
        for indx2,value2 in enumerate(datos):
            if value == datos[indx2]["producto"]:
                    lista_total_venta_producto[indx] += int(datos[indx2]["precio"])

    return lista_total_venta_producto


def precio_promedio(datos):

    listaDePreciosPromedio = []
    productos = obtener_productos_sin_repetir(datos)
    ganancias = obtener_ganancia_total_ventas_x_producto(productos)
    cantidadesVendidas = obtener_cantidades_vendidas(productos)

    for indx,value in enumerate(productos):
         
         listaDePreciosPromedio.append(f"{productos[indx]} {float(ganancias[indx] / cantidadesVendidas[indx]):.3f}")
         
    return listaDePreciosPromedio



def listar_precios(datos):
    print(" PRECIO PROMEDIO POR PRODUCTO\n")

    print("------------------------------------------------------------------------")
    print("PRODUCTOS                            PRECIOS PROMEDIO")
    print("------------------------------------------------------------------------")
    for i in datos:
        print( i)
        print("________________________________________________________________________")

precios = precio_promedio(datos)
listar_precios(precios)
````

### Ejercicio 5: Ranking de productos

Realizar un listado de los 3 productos mas vendidos ordenados la cantidad de unidades vendidas (ordenadas de mayor a menor)


````python
def ranking_productos(datos, top=3):
    
    lista_productos_y_unidades_vendidas = unidades_vendidas(datos)
    lista_de_tres_productos_mas_vendidios = []

    i = 0
    j = 1
    contador = 0
    longitud = len(lista_productos_y_unidades_vendidas[1])

    while contador < longitud:
        while j < longitud:
            if lista_productos_y_unidades_vendidas[1][i] < lista_productos_y_unidades_vendidas[1][j]:

                auxiliarUnidades = lista_productos_y_unidades_vendidas[1][i]
                auxiliarProductos = lista_productos_y_unidades_vendidas[0][i]
                lista_productos_y_unidades_vendidas[1][i] = lista_productos_y_unidades_vendidas[1][j]
                lista_productos_y_unidades_vendidas[0][i] = lista_productos_y_unidades_vendidas[0][j]
                lista_productos_y_unidades_vendidas[1][j] = auxiliarUnidades
                lista_productos_y_unidades_vendidas[0][j] = auxiliarProductos
            i += 1
            j += 1

        contador += 1
        i = 0
        j = 1

    contador = 0

    for indx,value in enumerate(lista_productos_y_unidades_vendidas[0]):

        lista_de_tres_productos_mas_vendidios.append((lista_productos_y_unidades_vendidas[0][indx], lista_productos_y_unidades_vendidas[1][indx]))
        contador += 1

        if contador == top:
            break


    return lista_de_tres_productos_mas_vendidios

def listar_ranking(ranking):
    print(" TOP 3 PRODUCTOS MAS VENDIDOS \n")
    print("----------------------------------------------------------")
    print("PRODUCTOS                            CANTIDADES VENDIDAS")
    print("----------------------------------------------------------")

    for indx, value in enumerate(ranking):
        print(f"{ranking[indx][0]} {ranking[indx][1]}")
        print("__________________________________________________________")


ranking = ranking_productos(datos)
listar_ranking(ranking)
````

### Ejercicio 6: Listar las ventas por mes

Realizar un listado del total de unidades vendidas por productos separada por mes.

````python

def ventas_por_mes(datos):
    meses = ['01','02','03','04','05','06','07','08','09','10','11','12']
    productos = obtener_productos_sin_repetir(datos)
    cantidadVendidaEnElMes = 0
    resultados = []

    for valueMes in meses:
        for indxProducto, valueProducto in enumerate(productos):
            for indxDato, valueDato in enumerate(datos):
                if(datos[indxDato]["fecha"][5:7] == valueMes and datos[indxDato]["producto"] == valueProducto):
                    cantidadVendidaEnElMes += int(datos[indxDato]["cantidad"])
            resultados.append(f"Producto:       {valueProducto}   |    Mes: {valueMes}   |    Cantidad vendida: {cantidadVendidaEnElMes}")
            cantidadVendidaEnElMes = 0
        resultados.append("")
    return resultados

def listar_ventas_mensuales(ventas):
    print("REGISTRO DE VENTAS MENSUALES \n")
    print("____________________________________________________________________________________________________________")
    for i in ventas:
        print(i)
        print("_________________________________________________________________________________________________________")

ventas = ventas_por_mes(datos)
listar_ventas_mensuales(ventas)
````

### Ejercicio 7: Informe general

Mostrar un listado de productos ordenados alfabeticamente que contengan el precio promedio, la cantidad de unidades vendidas y el importe total vendido para cada producto

````python
def resumen_ventas(datos):
    productos = obtener_productos_sin_repetir(datos)
    gananciaTotal = obtener_ganancia_total_ventas_x_producto(productos)
    cantidadesVendidas = obtener_cantidades_vendidas(productos)
    resumen = []

    for indx, value in enumerate(productos):
        resumen.append(f"{productos[indx]}{gananciaTotal[indx] / cantidadesVendidas[indx]:.3f}                          {cantidadesVendidas[indx]}                       {gananciaTotal[indx]}")

    return sorted(resumen)


def informe_ventas(datos):
    print("     INFORME GENERAL \n")
    print("------------------------------------------------------------------------------------------------------------------")
    print("PRODUCTO                            PRECIO PROMEDIO                  CANTIDAD VENDIDA         GANANCIA TOTAL")
    print("------------------------------------------------------------------------------------------------------------------")
    print("\n")
    for i in datos:
        print(i)
        print("________________________________________________________________________________________________________________")
        print("\n")


resumen = resumen_ventas(datos)
informe_ventas(resumen)
````
