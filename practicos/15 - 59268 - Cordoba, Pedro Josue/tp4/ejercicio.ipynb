# Simulación del Truco

Las **cartas** deben registra el numero, palo y valor de la carta. 
El valor debe reflejar la jerarquía de las cartas en el juego del truco.

```python
def calcular_valor(numero, palo):
    
    
    #lista de diccionarios con clave de tipo tupla
    jerarquia = [
            {(1,'espada'):12},
            {(1,'basto'):11},
            {(7,'espada'):10},
            {(7,'oro'):9},
            {(3,''):8},
            {(2,''):7},
            {(1,'copa'):6},
            {(1,'oro'):6},
            {(12,''):5},
            {(11,''):4},
            {(10,''):3},
            {(7,'copa'):2},
            {(7,'basto'):2},
            {(6,''):1},
            {(5,''):1},
            {(4,''):1}]
    
    posibleValor = 0
    
    for _ in jerarquia:
        for (num,pal),valor in _.items():
            
            if num == numero:
                posibleValor = valor 
                
            if num == numero and pal == palo:
                return valor
                
    return posibleValor
             
         
class Carta:

    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
        self.valor = calcular_valor(numero,palo) 

    def __str__(self):
        return str(self.valor)

c = Carta(12, "oro")
print(c.valor)
```

**Mazo**: Representan las cartas españolas de 1 al 12, en los palos de oro, copa, espada y basto.

Debe existir una función que permita barajar las cartas.
y otra función que permita repartir las cartas a los jugadores.

```python
import random

def obtener_cartas_del_truco():
    cartas = []
    palos = ['copa','oro','espada','basto']
    numeros = [1,2,3,4,5,6,7,10,11,12]

    for palo in palos:
        for numero in numeros:
            cartas.append((numero,palo))
    return cartas


def obtener_cantidad_cartas(pCartas):
    return len(pCartas)

class Mazo:
    def __init__(self):
        self.cartas = obtener_cartas_del_truco()
        self.cantidadCartas = obtener_cantidad_cartas(self.cartas)


    def mezclar(self):
        i = 0
        j = 0
        k = 1

        for contador in self.cartas:
            while k < self.cantidadCartas:
                i = random.randint(0,self.cantidadCartas - 1)
                j = random.randint(0,self.cantidadCartas - 1)
                
                if i > j:
                    auxiliar = self.cartas[i]
                    self.cartas[i] = self.cartas[j]
                    self.cartas[j] = auxiliar
                k+= 1


    def repartir(self, cantidad):
        cartas = []
        contador = 0

        if len(self.cartas) >= cantidad:
            while contador < cantidad:
                cartas.append(self.cartas.pop())
                contador += 1
        else:
            print("Sin cartas en el mazo")
            
        return cartas
    
    #Esta funcion nos permite regresar al objeto al estado inicial
    def reset(self):
        self.__init__()

    def __len__(self):
        return len(self.cartas)

m = Mazo()
m.mezclar()
print(m.repartir(3))
```

**Jugador**: Deben tener un nombre y las cartas que tiene en la mano. Debe acumular los puntos ganados.

Nota: Para simplificar el problema, el jugador elige la carta a jugar de forma aleatoria.

```python
class Jugador:
    pass
    def __init__(self, nombre):
        self.nombre = nombre

    def recibir_cartas(self, cartas):
        self.cartas = cartas
    
    def jugar_carta(self):

        if len(self.cartas) != 0:
             #retornamos una carta aleatoria entre 0 y la cantidad de cartas que queden
             return self.cartas.pop(random.randint(0,len(self.cartas) - 1))
        
    def __str__(self):
        return str(self.cartas)

j = Jugador("Juan")
j.recibir_cartas(m.repartir(3))
print(j.jugar_carta())
```
