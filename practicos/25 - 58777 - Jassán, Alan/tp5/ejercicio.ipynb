## Productos
- `codigo`: 4 digitos
- `nombre`: 1 a 100 caracteres
- `precio`: 10 a 10000
- `tipo`: 0 a 20 caracteres
- `cantidad`: 0 a 100 
- `valorTotal`: cantidad * precio

```python
class Producto:
    def __init__(self, codigo, nombre, precio, tipo, cantidad):        
        self.codigo = codigo
        self.nombre = nombre
        self.precio = precio
        self.tipo = tipo
        self.cantidad = cantidad

    def validar_codigo(self, codigo):
        if len(codigo) != 4 or not codigo.isdigit():
            raise ValueError("El codigo debe ser un numero de 4 digitos")
        return codigo
    
    def validar_nombre(self, nombre):
        if not (1 <= len(nombre) <= 100):
            raise ValueError("El nombre debe tener entre 1 y 100 caracteres")
        return nombre
    
    def validar_precio(self, precio):
        if not (10 <= precio <= 10000):
            raise ValueError("El precio debe estar entre 10 y 10000")
        return precio
    
    def validar_tipo(self, tipo):
        if not (0 <= len(tipo) <= 20):
            raise ValueError("El tipo debe tener entre 0 y 20 caracteres")
        return tipo
    
    def validar_cantidad(self, cantidad):
        if not (0 <= cantidad <= 100):
            raise ValueError("La cantidad debe estar entre 0 y 100")
        return cantidad
    
    def valorTotal(self):
        return self.precio * self.cantidad

    # Getters y setters
    @property
    def codigo(self):
        return self._codigo
    
    @codigo.setter
    def codigo(self, codigo):
        try:
            self._codigo = self.validar_codigo(codigo)
        except ValueError as e:
            print(f"Error al asignar el codigo: {e} ")
    
    @property
    def nombre(self):
        return self._nombre
    
    @nombre.setter
    def nombre(self, nombre):
        try:
            self._nombre = self.validar_nombre(nombre)
        except ValueError as e:
            print(f"Error al asignar el nombre: {e}")

    @property
    def precio(self):
        return self._precio
    
    @precio.setter
    def precio(self, precio):
        try:
            self._precio = self.validar_precio(precio)
        except ValueError as e:
            print(f"Error al asignar el precio: {e}")
    
    @property
    def tipo(self):
        return self._tipo
    
    @tipo.setter
    def tipo(self, tipo):
        try:
            self._tipo = self.validar_tipo(tipo)
        except ValueError as e:
            print(f"Error al asignar el tipo: {e}")

    @property
    def cantidad(self):
        return self._cantidad
    
    @cantidad.setter
    def cantidad(self, cantidad):
        try:
            self._cantidad = self.validar_cantidad(cantidad)
        except ValueError as e:
            print(f"Error al asignar la cantidad: {e}")

    def valorTotal(self):
        return self.precio * self.cantidad
```

# Ofertas
Debe permitir aplicar oferctas a codigos espeficicos de productos y a tipos de productos
- `descripcion`: 1 a 100 caracteres
- `codigos`: lista de codigos de productos
- `tipos`: lista de tipo de producto
- `esAplicable(producto, cantidad)`: retorna si la oferta es aplicable a ese producto
- `aplicar(producto, cantidad)`: retorna el precio final del producto con la oferta aplicada


```python
class Oferta:
    def __init__(self, descripcion = None, codigos = [], tipos = []):
        self.descripcion = self.validar_descripcion(descripcion)
        self.codigos = codigos if codigos else []
        self.tipos = tipos if tipos else []

    def validar_descripcion(self, descripcion):
        if descripcion is not None and not (1 <= len(descripcion) <= 100):
            raise ValueError("La descripcion debe tener entre 1 y 100 caracteres")
        return descripcion

    def esAplicable(self, producto):
        return (producto.codigo in self.codigos) or (producto.tipo in self.tipos)
    
    def aplicar(self, producto, cantidad):
        return producto.precio * cantidad

    def calcularDescuento(self, producto, cantidad):
        return 0

class OfertaDescuento(Oferta):
    def __init__(self, descuento, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.descuento = self.validar_descuento(descuento)
    
    def validar_descuento(self, descuento):
        if not (0 <= descuento <= 100):
            raise ValueError("El descuento debe estar entre 0 y 100")
        return descuento
    
    def calcularDescuento(self, producto, cantidad):
        if self.esAplicable(producto):
            descuento_total = producto.precio * (self.descuento / 100) * cantidad
            return descuento_total
        return 0
    
class Oferta2x1(Oferta):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)

    def calcularDescuento(self, producto, cantidad):
        if self.esAplicable(producto):
            descuento = producto.precio * (cantidad // 2)
            return descuento
        return 0
```

# Catalogo
- `leer(archivo) `    : Carga los productos desde el archivo
- `guardar(archivo)`  : Guarda los productos en el archivo
- `agregar(producto)` : Agrega un producto al catalogo
- `buscar(codigo)`    : Busca un producto por codigo o None si no existe
- `registrarOferta(oferta)`  : Registra una oferta
- `buscarOferta(producto, cantidad)`: Busca una oferta por codigo o None si no existe
- `calcularDescuento(producto, cantidad)`: Calcula el descuento de una oferta
- `cantidadProductos`: Retorna la cantidad de productos en el catalogo
- `cantidadUnidades`: Retorna la cantidad de unidades en el catalogo
- `valorTotal`: retorna el valor total del catalogo sin descuentos
- `informe()`: retorna un string con el informe del catalogo 

```python
class Catalogo:
    def __init__(self):
        self.productos = []
        self.ofertas = []
        
    @classmethod
    def leer(cls, archivo):
        catalogo = cls()
        with open(archivo, 'r') as file:
            contador = 0
            for line in file:
                # Saltar la primera linea del archivo
                if contador == 0:
                    contador += 1
                    continue
                codigo, nombre, precio, tipo, cantidad = line.strip().split(',')
                catalogo.agregar(Producto(codigo, nombre, int(precio), tipo, int(cantidad)))
        return catalogo

    def guardar(self, archivo):
        with open(archivo, 'w') as file:
            file.write('codigo,nombre,precio,tipo,cantidad\n')
            for producto in self.productos:
                file.write(f'{producto.codigo},{producto.nombre},{producto.precio},{producto.tipo},{producto.cantidad}\n')
    
    def agregar(self, *productos):
            productos = list(productos)
            for producto in productos:
                self.productos.append(producto)

    def buscar(self, codigo):
        for producto in self.productos:
            if producto.codigo == codigo:
                return producto
        return None
    
    def registrarOferta(self, oferta):
        self.ofertas.append(oferta)

    def buscarOferta(self, producto, cantidad):
        for oferta in self.ofertas:
            if oferta.esAplicable(producto) and oferta.calcularDescuento(producto, cantidad) > 0:
                return oferta
        return None
    
    def calcularDescuento(self, producto, cantidad):
        oferta = self.buscarOferta(producto, cantidad)
        if oferta:
            return oferta.calcularDescuento(producto, cantidad)
        return 0
    
    def vender(self, producto, cantidad):
        for i in range(len(self.productos)):
            if self.productos[i].codigo == producto.codigo:
                if cantidad <= self.productos[i].cantidad:
                    self.productos[i].cantidad -= cantidad
                    if self.productos[i].cantidad == 0:
                        del self.productos[i]
                    return True
                else:
                    print(f"Stock de {self.productos[i].nombre} insuficiente")
                    return False
        print(f"Producto {producto.codigo} no encontrado")
        return False
    

        # encontrar_producto = self.buscar(producto.codigo)
        # if encontrar_producto:
        #     if cantidad <= encontrar_producto.cantidad:
        #         encontrar_producto.cantidad -= cantidad
        #         if encontrar_producto.cantidad == 0:
        #             self.productos.remove(encontrar_producto)
        #         return True
        #     else:
        #         print(f"Stock de {producto.nombre} insuficiente")
        #         return False
        # else:
        #     print(f"Producto {producto.codigo} no encontrado")
        #     return False

    @property
    def cantidadProductos(self):
        return len(self.productos)
    
    @property
    def cantidadUnidades(self):
        return sum(producto.cantidad for producto in self.productos)
    
    @property
    def valorTotal(self):
        return sum(producto.valorTotal() for producto in self.productos)
    
    def informe(self):
        informe = f"Cantidad de productos: {self.cantidadProductos}\n"
        informe += f"Cantidad de unidades: {self.cantidadUnidades}\n"

        precio_promedio = self.valorTotal / self.cantidadProductos if self.cantidadProductos else 0
        informe += f"Precio Promedio: ${precio_promedio:.2f}\n"
        informe += f"Valor total: ${self.valorTotal}\n"
        informe += f"Tipos de productos: \n"
        
        tipos_productos = {}
        for producto in self.productos:
            if producto.tipo in tipos_productos:
                tipos_productos[producto.tipo] += producto.cantidad
            else:
                tipos_productos[producto.tipo] = producto.cantidad

        for tipo, cantidad in tipos_productos.items():
            informe += f"{tipo}: {cantidad}\n"

        informe += f"Ofertas: \n"
        for oferta in self.ofertas:
            informe += f"{oferta.descripcion}\n"

        informe += f"Oferta 2x1: \n"
        # Itero para verificar la instancia de Oferta2x1, True agrega la lista
        ofertas2x1 = [oferta for oferta in self.ofertas if isinstance(oferta, Oferta2x1)]
        for oferta in ofertas2x1:
            informe += "  - Oferta 2x1\n"

        return informe
```

```python
## NO MODIFIQUE ESTE CODIGO ##

# Prueba del catálogo 

catalogo = Catalogo()
p1 = Producto('0001', 'Coca Cola',  1500, 'gaseosa', 10)
p2 = Producto('0002', 'Pepsi Cola', 1200, 'gaseosa', 20)
p3 = Producto('0003', 'Sonrisa',    1200, 'galleta', 30)
p4 = Producto('0004', 'Oreo',       2300, 'galleta', 40)

## Agregar productos al catalogo 
catalogo.agregar(p1)
catalogo.agregar(p2)
catalogo.agregar(p3)
catalogo.agregar(p4)

assert catalogo.cantidadProductos == 4
assert catalogo.cantidadUnidades == 100

assert catalogo.valorTotal == 167000

## Calcular descuentos segun las ofertas registradas
assert catalogo.calcularDescuento(p1, 5) == 0
assert catalogo.calcularDescuento(p2, 5) == 0

# Ofertas no acumulables 
catalogo.registrarOferta(Oferta2x1(tipos=['galleta']))
catalogo.registrarOferta(OfertaDescuento(10, codigos=['0001', '0003']))

assert catalogo.calcularDescuento(p1, 5) == 750
assert catalogo.calcularDescuento(p2, 5) == 0
assert catalogo.calcularDescuento(p3, 5) == 2400

assert catalogo.valorTotal == 167000.0
catalogo.guardar('catalogo-prueba.csv') ## Guardar datos antes de vender

# Vender afecta la cantidad de unidades y el valor total
catalogo.vender(p3, 3)   

# Verificar que el informe se genere correctamente

informe = catalogo.informe()
assert "Cantidad de productos: " in informe
assert "Cantidad de unidades: " in informe
assert "Precio Promedio: " in informe
assert "Valor total: " in informe
assert "Tipos de productos: " in informe
assert "gaseosa" in informe
assert "galleta" in informe
assert "Ofertas:" in informe 
assert "Oferta 2x1" in informe
assert catalogo.cantidadUnidades == 97
assert catalogo.valorTotal == 163400

# Buscar por código
assert catalogo.buscar('0001') == p1
assert catalogo.buscar('0002') == p2
assert catalogo.buscar('0099') is None 

# Recuperar los datos guardados  
c2 = Catalogo.leer('catalogo-prueba.csv')

assert c2.cantidadProductos == 4
assert c2.cantidadUnidades == 100

# Valor antes de guardar
assert c2.valorTotal == 167000.0

print("Prueba pasada exitosamente!")
```

# Cliente
- `nombre`: 1 a 100 caracteres
- `cuit`: 13 digitos (formato XX-XXXXXXXX-X)

```python
class Cliente: 
    def __init__(self, nombre, cuit):
        self.nombre = nombre
        self.cuit = cuit

    def validar_nombre(self, nombre):
        if not (1 <= len(nombre) <= 100):
            raise ValueError("El nombre tener entre 1 y 100 caracteres")
        return nombre
    
    def validar_cuit(self, cuit):
        if not (cuit[0:2].isdigit() and cuit[2] == '-' and cuit[3:11].isdigit() and cuit[11] == '-' and cuit[12].isdigit()):
            raise ValueError("El CUIT debe tener el formato XX-XXXXXXXX-X")
        return cuit
    
    @property
    def nombre(self):
        return self._nombre
    
    @nombre.setter
    def nombre(self, nombre):
        try:
            self._nombre = self.validar_nombre(nombre)
        except ValueError as e:
            print(f"Error al asignar el nombre: {e} ")

    @property
    def cuit(self):
        return self._cuit
    
    @cuit.setter
    def cuit(self, cuit):
        try:
            self._cuit = self.validar_cuit(cuit)
        except ValueError as e:
            print(f"Error al asignar el CUIT: {e} ")
    
```

```python
class Factura:
    _ultimo_num = 0
    def __init__(self, catalogo, cliente):
        self.numero = Factura.nuevoNumero()
        self.catalogo = catalogo
        self.cliente = cliente
        self.productos = {}

    @classmethod
    def ultimaFactura(cls, numero):
        cls._ultimo_num = numero

    @classmethod
    def nuevoNumero(cls):
        cls._ultimo_num += 1
        return cls._ultimo_num
    
    def agregar(self, producto, cantidad):
        if cantidad <= 0:
            raise ValueError("La cantidad debe ser mayor a cero")
        # Buscar el producto original en el catálogo
        for product in self.catalogo.productos:
            if product.codigo == producto.codigo:
                product.cantidad -= cantidad # Actualizar
                break
        # Agregar el produtco en la factura
        if producto.codigo in self.productos:
            self.productos[producto.codigo] += cantidad
        else:
            self.productos[producto.codigo] = cantidad

    @property
    def cantidadProductos(self):
        return len(self.productos)
    
    @property
    def cantidadUnidades(self):
        return sum(self.productos.values())
    
    @property
    def subtotal(self):
        subtotal = 0
        for codigo, cantidad in self.productos.items():
            producto = self.catalogo.buscar(codigo)
            subtotal += producto.precio * cantidad
        return subtotal
    
    @property
    def descuentos(self):
        descuento = 0
        for codigo, cantidad in self.productos.items():
            producto = self.catalogo.buscar(codigo)
            descuento += self.catalogo.calcularDescuento(producto, cantidad)
        return descuento
    
    @property
    def total(self):
        return self.subtotal - self.descuentos
    
    def imprimir(self):
        impresion = f"Factura Número: {self.numero}\n"
        impresion += f"Cliente: {self.cliente.nombre}\n"

        impresion += f"Productos:\n"
        for codigo, cantidad in self.productos.items():
            producto = self.catalogo.buscar(codigo)
            impresion += f"{cantidad} {producto.nombre} - ${producto.precio}\n"
            oferta = self.catalogo.buscarOferta(producto, cantidad)
            if oferta:
                if isinstance(oferta, OfertaDescuento):
                    impresion += f"Descuento {oferta.descuento}%\n"
                elif isinstance(oferta, Oferta2x1):
                    impresion += f"Oferta 2x1\n"
                # else:
                #     return 0
        impresion += f"TOTAL:\n"
        impresion += f"{self.total:.2f}\n"

        return impresion
```
