# Simulación del Truco

Las **cartas** deben registra el numero, palo y valor de la carta. 
El valor debe reflejar la jerarquía de las cartas en el juego del truco.

```python
import random

def calcularvalores(numero, palo):
    jerarquia = {
        (1, "oro"): 14,
        (1, "copa"): 13,
        (1, "espada"): 12,
        (1, "basto"): 11,
        (2, ): 10,
        (3, ): 9,
        (4, ): 8,
        (5, ): 7,
        (6, ): 6,
        (7, ): 5,
        (10, ): 4,
        (11, ): 3,
        (12, ): 2,
    }
    return jerarquia.get((numero, palo), jerarquia.get((numero,), 0))

class Cartas:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
        self.valor = calcularvalores(numero, palo)

    def __repr__(self):
        return f"{self.numero} de {self.palo}"

# Crear una carta y mostrar su representación
c = Cartas(1, "oro")
print(c)

# Ejemplo de lista de cartas
cartas = [Cartas(5, "espada"), Cartas(6, "espada"), Cartas(10, "espada")]
print(cartas)
```

**Mazo**: Representan las cartas españolas de 1 al 12, en los palos de oro, copa, espada y basto.

Debe existir una función que permita barajar las cartas.
y otra función que permita repartir las cartas a los jugadores.

```python
import random

numeross = [1, 2, 3, 4, 5, 6, 7, 10, 11, 12]
palos = ["oro", "copa", "espada", "basto"]

class Cartas:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
        self.valor = self.calcular_valorres()

    def calcular_valorres(self):
        jerarquia = {
            (1, "oro"): 14,
            (1, "copa"): 13,
            (1, "espada"): 12,
            (1, "basto"): 11,
            (2, ): 10,
            (3, ): 9,
            (4, ): 8,
            (5, ): 7,
            (6, ): 6,
            (7, ): 5,
            (10, ): 4,
            (11, ): 3,
            (12, ): 2,
        }
        return jerarquia.get((self.numero, self.palo), jerarquia.get((self.numero,), 0))

    def __repr__(self):
        return f"{self.numero} de {self.palo}"

class Mazo:
    def __init__(self):
        self.cartas = [Cartas(numero, palo) for numero in numeross for palo in palos]

    def mezclar(self):
        random.shuffle(self.cartas)

    def repartir(self, cantidad):
        return [self.cartas.pop() for _ in range(cantidad)]

# Crear y mezclar el mazo
m = Mazo()
m.mezclar()

# Repartir cartas y mostrar la salida
cartas_repartidas = m.repartir(3)
print(f"Juan recibió {cartas_repartidas} y tiene {sum(carta.valor for carta in cartas_repartidas)} puntos")
print(cartas_repartidas[2])
```

**Jugador**: Deben tener un nombre y las cartas que tiene en la mano. Debe acumular los puntos ganados.

**Partida**: Debe tener dos jugadores, un mazo de cartas y un puntaje. Debe permitir jugar una carta por ronda y definir el ganador de la ronda. Debe permitir jugar hasta que se terminen las cartas del mazo.

```python
import random

numeross = [1, 2, 3, 4, 5, 6, 7, 10, 11, 12]
palos = ["oro", "copa", "espada", "basto"]

class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
        self.valor = self.calcular_valor()

    def calcular_valor(self):
        jerarquia = {
            (1, "oro"): 14,
            (1, "copa"): 13,
            (1, "espada"): 12,
            (1, "basto"): 11,
            (2, ): 10,
            (3, ): 9,
            (4, ): 8,
            (5, ): 7,
            (6, ): 6,
            (7, ): 5,
            (10, ): 4,
            (11, ): 3,
            (12, ): 2,
        }
        return jerarquia.get((self.numero, self.palo), jerarquia.get((self.numero,), 0))

    def __repr__(self):
        return f"{self.numero} de {self.palo}"

class Mazo:
    def __init__(self):
        self.cartas = [Carta(numeross, palo) for numeross in numeross for palo in palos]

    def mezclar(self):
        random.shuffle(self.cartas)

    def repartir(self, cantidad):
        return [self.cartas.pop() for _ in range(cantidad)]

class Jugador:
    def __init__(self, nombre):
        self.nombre = nombre
        self.puntos = 0
        self.cartas = []

    def recibir_cartas(self, cartas):
        self.cartas = cartas
        puntos = sum(carta.valor for carta in cartas)
        print(f"{self.nombre} recibió {cartas} y posee {puntos} puntos")

    def jugar_carta(self):
        return self.cartas.pop()

class Partida:
    def __init__(self, jugador1, jugador2):
        self.jugador1 = jugador1
        self.jugador2 = jugador2
        self.rondas = 0

    def jugar_ronda(self):
        valor_jugador1 = sum(carta.valor for carta in self.jugador1.cartas)
        valor_jugador2 = sum(carta.valor for carta in self.jugador2.cartas)

        if valor_jugador1 > valor_jugador2:
            return self.jugador1
        elif valor_jugador2 > valor_jugador1:
            return self.jugador2
        else:
            return None

    def jugar(self):
        while self.rondas < 3 and all([self.jugador1.puntos < 15, self.jugador2.puntos < 15]):
            m = Mazo()
            m.mezclar()
            self.jugador1.recibir_cartas(m.repartir(3))
            self.jugador2.recibir_cartas(m.repartir(3))

            ganador = self.jugar_ronda()

            if ganador:
                ganador.puntos += 1
                print(f"La persona que vano la gane la ronde eeeeeees: {ganador.nombre}")
            else:
                print("Empate")

            self.rondas += 1
            print(f"Rondas jugadas: {self.rondas}")
            print(f"Puntos de {self.jugador1.nombre}: {self.jugador1.puntos}")
            print(f"Puntos de {self.jugador2.nombre}: {self.jugador2.puntos}")

# Crear y jugar una partida
p = Partida(Jugador("Juan"), Jugador("Pedro"))
p.jugar()
```
