# Simulación del Truco

Las **cartas** deben registra el numero, palo y valor de la carta. 
El valor debe reflejar la jerarquía de las cartas en el juego del truco.

```python
def calcular_valor(numero, palo):
    jerarquia_especial = {
        (1, "espadas"): 31,
        (1, "bastos"): 30,
        (7, "espadas"): 29,
        (7, "oros"): 28,
        3: 27,
        2: 26,
        (1, "copas"): 25,
        (1, "oros"): 25,
        "rey": 24,
        "caballo": 23,
        "sota": 22,
        (7, "copas"): 21,
        (7, "bastos"): 21,
        6: 20,
        5: 19,
        4: 18
    }
    
    if (numero, palo) in jerarquia_especial:
        return jerarquia_especial[(numero, palo)]
    elif numero in jerarquia_especial:
        return jerarquia_especial[numero]
    else:
        return 0 

class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
        self.valor = calcular_valor(numero, palo)
    
    def __str__(self):
        return f"{self.numero} de {self.palo} (valor: {self.valor})"

c = Carta(1, "oros")
print(c)
```

**Mazo**: Representan las cartas españolas de 1 al 12, en los palos de oro, copa, espada y basto.

Debe existir una función que permita barajar las cartas.
y otra función que permita repartir las cartas a los jugadores.

```python
import random

class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
    
    def __str__(self):
        return f"{self.numero} de {self.palo}"

class Mazo:
    def __init__(self):
        self.cartas = [Carta(numero, palo) for numero in range(1, 13) for palo in ["oro", "copa", "espada", "basto"]]
    
    def mezclar(self):
        random.shuffle(self.cartas)
    
    def repartir(self, cantidad):
        if cantidad <= len(self.cartas):
            mano = self.cartas[:cantidad]
            self.cartas = self.cartas[cantidad:]
            return mano
        else:
            return [] 
    
    def __str__(self):
        return ', '.join(str(carta) for carta in self.cartas)

m = Mazo() 
m.mezclar()  

cartas_repartidas = m.repartir(3)
for carta in cartas_repartidas:
    print(carta)
```

**Jugador**: Deben tener un nombre y las cartas que tiene en la mano. Debe acumular los puntos ganados.

Nota: Para simplificar el problema, el jugador elige la carta a jugar de forma aleatoria.

```python
import random

class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
    
    def __str__(self):
        return f"{self.numero} de {self.palo}"

class Jugador:
    def __init__(self, nombre):
        self.nombre = nombre
        self.mano = []  
        self.puntos = 0  
    
    def recibir_cartas(self, cartas):
        self.mano.extend(cartas)
    
    def jugar_carta(self):
        if self.mano:
            carta_a_jugar = random.choice(self.mano)
            self.mano.remove(carta_a_jugar)
            return carta_a_jugar
        else:
            return None  
    
    def acumular_puntos(self, puntos):
        self.puntos += puntos
    
    def __str__(self):
        return f"Jugador: {self.nombre}, Cartas en mano: {[str(carta) for carta in self.mano]}, Puntos: {self.puntos}"

class Mazo:
    def __init__(self):
        self.cartas = [Carta(numero, palo) for numero in range(1, 13) for palo in ["oro", "copa", "espada", "basto"]]
    
    def mezclar(self):
        random.shuffle(self.cartas)
    
    def repartir(self, cantidad):
        if cantidad <= len(self.cartas):
            mano = self.cartas[:cantidad]
            self.cartas = self.cartas[cantidad:]
            return mano
        else:
            return []  

m = Mazo()  
m.mezclar() 

j = Jugador("Juan")  
j.recibir_cartas(m.repartir(3)) 

print(f"Antes de jugar: {j}")
print(f"Juan juega: {j.jugar_carta()}")
print(f"Después de jugar: {j}")
```

**Partida**: Debe tener dos jugadores, un mazo de cartas y un puntaje. Debe permitir jugar una carta por ronda y definir el ganador de la ronda. Debe permitir jugar hasta que se terminen las cartas del mazo.

```python
import random
class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
    
    def __str__(self):
        return f"{self.numero} de {self.palo}"

class Jugador:
    def __init__(self, nombre):
        self.nombre = nombre
        self.mano = []  
        self.puntos = 0  
    
    def recibir_cartas(self, cartas):
        self.mano.extend(cartas)
    
    def jugar_carta(self):
        if self.mano:
            carta_a_jugar = random.choice(self.mano)
            self.mano.remove(carta_a_jugar)
            return carta_a_jugar
        else:
            return None 
    
    def acumular_puntos(self, puntos):
        self.puntos += puntos
    
    def __str__(self):
        return f"Jugador: {self.nombre}, Cartas en mano: {[str(carta) for carta in self.mano]}, Puntos: {self.puntos}"

class Mazo:
    def __init__(self):
        self.cartas = [Carta(numero, palo) for numero in range(1, 13) for palo in ["oro", "copa", "espada", "basto"]]
    
    def mezclar(self):
        random.shuffle(self.cartas)
    
    def repartir(self, cantidad):
        if cantidad <= len(self.cartas):
            mano = self.cartas[:cantidad]
            self.cartas = self.cartas[cantidad:]
            return mano
        else:
            return []  
    
    def quedan_cartas(self):
        return len(self.cartas) > 0 

class Partida:
    def __init__(self, jugador1, jugador2):
        self.jugador1 = jugador1
        self.jugador2 = jugador2
        self.mazo = Mazo()  
        self.puntaje_maximo = 15 
    
    def jugar_ronda(self):
        carta_jugador1 = self.jugador1.jugar_carta()
        carta_jugador2 = self.jugador2.jugar_carta()

        if carta_jugador1 and carta_jugador2:
            print(f"{self.jugador1.nombre} juega: {carta_jugador1}")
            print(f"{self.jugador2.nombre} juega: {carta_jugador2}")
            
            if carta_jugador1.numero > carta_jugador2.numero:
                print(f"¡{self.jugador1.nombre} gana la ronda!")
                self.jugador1.acumular_puntos(1)
            elif carta_jugador2.numero > carta_jugador1.numero:
                print(f"¡{self.jugador2.nombre} gana la ronda!")
                self.jugador2.acumular_puntos(1)
            else:
                print("¡Empate en la ronda!")
        else:
            print("Uno de los jugadores no tiene cartas.")

    def jugar(self):
        self.mazo.mezclar()
        
        while self.jugador1.puntos < self.puntaje_maximo and self.jugador2.puntos < self.puntaje_maximo and self.mazo.quedan_cartas():
            
            self.jugador1.recibir_cartas(self.mazo.repartir(3))
            self.jugador2.recibir_cartas(self.mazo.repartir(3))

            for _ in range(3):
                self.jugar_ronda()
            
            print(f"Puntaje: {self.jugador1.nombre} {self.jugador1.puntos} - {self.jugador2.nombre} {self.jugador2.puntos}")
        
        if self.jugador1.puntos >= self.puntaje_maximo:
            print(f"¡{self.jugador1.nombre} ha ganado la partida!")
        elif self.jugador2.puntos >= self.puntaje_maximo:
            print(f"¡{self.jugador2.nombre} ha ganado la partida!")
        else:
            print("Se acabaron las cartas, la partida ha terminado sin un ganador claro.")

p = Partida(Jugador("Juan"), Jugador("Pedro"))
p.jugar()
```
