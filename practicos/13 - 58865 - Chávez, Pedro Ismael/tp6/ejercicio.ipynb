### Ejercicio 1: Cargar Datos de ventas.

El archivo datos.dat contiene el registro de las ventas realizadas.

Tiene un formato de ancho fijo:
- `fecha`:    10 lugares
- `producto`: 30 lugares
- `precio`:   10 lugares
- `cantidad`:  5 lugares

Hacer una funcion que cargue los datos en un DataFrame de Pandas.

```python
import pandas as pd

def cargar_datos():
    lineas = open("datos.dat", "r").readlines()
    datos = []

    for linea in lineas:
        fecha = linea[:10].strip()
        producto = linea[10:40].strip()
        precio = int(linea[40:50].strip())
        cantidad = int(linea[50:].strip())
        datos.append({'fecha': fecha, 'producto': producto, 'precio': precio, 'cantidad': cantidad})

    df = pd.DataFrame(datos)
    return df

datos = cargar_datos()
datos 
```

### Ejercicio 2: Calcular el total de ventas.
Hacer una funci√≥n que sume los importes vendidos (precio * cantidad) y las cantidades.


```python
import pandas as pd

def calcular_totales(datos):
    with open("datos.dat", "r") as file:
        lineas = file.readlines()


    precios = []
    cantidades = []


    for linea in lineas:
        precio = int(linea[40:50].strip())  # Extraer precio
        cantidad = int(linea[50:].strip())  # Extraer cantidad
        precios.append(precio)
        cantidades.append(cantidad)


    df = pd.DataFrame({
        'precio': precios,
        'cantidad': cantidades
    })


    df['importe'] = df['precio'] * df['cantidad']
    
    importe = df['importe'].sum()
    cantidad = df['cantidad'].sum()

    #print(df)
    return importe, cantidad


importe, cantidad = calcular_totales(datos)

print(f"Las ventas fueron de ${importe:.2f} en {cantidad} unidades")
```

### Ejercicio 5: Ranking de productos
Realizar un listado de los 3 productos m√°s vendidos ordenados por la cantidad de unidades vendidas (ordenadas de mayor a menor)


```python
class Ventas:
    def __init__(self, archivo):
        self.archivo = archivo
        self.datos = self.cargar_datos()

    def cargar_datos(self):
        lineas = open(self.archivo, "r").readlines()
        datos = []

        for linea in lineas:
            fecha = linea[:10].strip()
            producto = linea[10:40].strip()
            precio = int(linea[40:50].strip())
            cantidad = int(linea[50:].strip())
            datos.append({'fecha': fecha, 'producto': producto, 'precio': precio, 'cantidad': cantidad})

        return datos

    def ranking_productos(self, top=3):
        lineas = open(self.archivo, "r").readlines()
        productos = {}

        for linea in lineas:
            producto = linea[10:40].strip()
            cantidad = int(linea[50:].strip())

            if producto in productos:
                productos[producto] += cantidad
            else:
                productos[producto] = cantidad

        ranking = sorted(productos.items(), key=lambda x: x[1], reverse=True)[:top]

  
        df_ranking = pd.DataFrame(ranking, columns=['Producto', 'Cantidad Vendida'])

    
        print(df_ranking)

        return df_ranking

    def listar_ranking(self, ranking):
        for producto, cantidad in ranking.itertuples(index=False):
            print(f"\n{producto} {cantidad}")

ventas = Ventas("datos.dat")
ranking = ventas.ranking_productos()
ventas.listar_ranking(ranking)
```

### Ejercicio 6: Listar las ventas por mes
Realizar un listado del total de unidades vendidas por producto separado por mes.


```python
class Ventas:
    def __init__(self, archivo):
        self.archivo = archivo
        self.datos = self.cargar_datos()

    def cargar_datos(self):
        lineas = open(self.archivo, "r").readlines()
        datos = []

        for linea in lineas:
            fecha = linea[:10].strip()
            producto = linea[10:40].strip()
            precio = int(linea[40:50].strip())
            cantidad = int(linea[50:].strip())
            datos.append({'fecha': fecha, 'producto': producto, 'precio': precio, 'cantidad': cantidad})

        return datos

    def ventas_por_mes(self):
        datos = self.datos
        df = pd.DataFrame(datos)
        df['fecha'] = pd.to_datetime(df['fecha'], format='%Y-%m-%d')
        df['mes'] = df['fecha'].dt.to_period('M')
        ventas_mensuales = df.groupby(['mes', 'producto'])['cantidad'].sum().unstack().fillna(0)

        return ventas_mensuales

    def listar_ventas_mensuales(self, ventas_mensuales):
        print(ventas_mensuales)

ventas = Ventas("datos.dat")
ventas_mensuales = ventas.ventas_por_mes()
ventas.listar_ventas_mensuales(ventas_mensuales)
```

### Ejercicio 7: Informe general

Mostrar un listado de productos ordenados alfabeticamente que contengan el precio promedio, la cantidad de unidades vendidas y el importe total vendido para cada producto

```python
def calcular_totales(datos):
    total_importe = sum(d['precio'] * d['cantidad'] for d in datos)
    total_unidades = sum(d['cantidad'] for d in datos)
    return total_importe, total_unidades

def unidades_vendidas(datos):
    productos = {}
    for dato in datos:
        if dato['producto'] in productos:
            productos[dato['producto']] += dato['cantidad']
        else:
            productos[dato['producto']] = dato['cantidad']
    return productos

def ventas_por_mes(datos):
    ventas = {}
    for dato in datos:
        mes = dato['fecha'][5:7]
        if mes in ventas:
            ventas[mes] += dato['cantidad']
        else:
            ventas[mes] = dato['cantidad']
    return ventas

def ranking_productos(datos):
    productos = unidades_vendidas(datos)
    ranking = sorted(productos.items(), key=lambda x: x[1], reverse=True)
    return ranking

def precio_promedio(datos):
    productos = {}
    for dato in datos:
        if dato['producto'] in productos:
            productos[dato['producto']].append(dato['precio'])
        else:
            productos[dato['producto']] = [dato['precio']]
    promedios = {k: sum(v) / len(v) for k, v in productos.items()}
    return promedios

def listar_ventas(unidades):
    df_unidades = pd.DataFrame(list(unidades.items()), columns=['Producto', 'Cantidad'])
    print(df_unidades)

def listar_ventas_mensuales(ventas):
    df_ventas = pd.DataFrame(list(ventas.items()), columns=['Mes', 'Cantidad'])
    print(df_ventas)

def listar_ranking(ranking):
    df_ranking = pd.DataFrame(ranking, columns=['Producto', 'Cantidad'])
    print(df_ranking)

def listar_precios(precios):
    df_precios = pd.DataFrame(list(precios.items()), columns=['Producto', 'Precio Promedio'])
    print(df_precios)

class Ventas:
    def __init__(self, archivo):
        self.archivo = archivo
        self.datos = self.cargar_datos()

    def cargar_datos(self):
        lineas = open(self.archivo, "r").readlines()
        datos = []
        for linea in lineas:
            fecha = linea[:10].strip()
            producto = linea[10:40].strip()
            precio = int(linea[40:50].strip())
            cantidad = int(linea[50:].strip())
            datos.append({'fecha': fecha, 'producto': producto, 'precio': precio, 'cantidad': cantidad})
        return datos

    def resumen_ventas(self, datos):
        lineas = open(self.archivo, "r").readlines()
        datos = []
        for linea in lineas:
            fecha = linea[:10].strip()
            producto = linea[10:40].strip()
            precio = int(linea[40:50].strip())
            cantidad = int(linea[50:].strip())
            datos.append({'fecha': fecha, 'producto': producto, 'precio': precio, 'cantidad': cantidad})

        resumen = {}
        resumen['totales'] = calcular_totales(datos)
        resumen['unidades'] = unidades_vendidas(datos)
        resumen['ventas'] = ventas_por_mes(datos)
        resumen['ranking'] = ranking_productos(datos)
        resumen['precios'] = precio_promedio(datos)
        return resumen

    def informe_ventas(self, datos):
        print(f"Ventas totales: ${datos['totales'][0]:.2f} en {datos['totales'][1]} unidades\n")
        print("Unidades vendidas por producto:")
        listar_ventas(datos['unidades'])
        print("\n")
        print("Ventas por mes:")
        listar_ventas_mensuales(datos['ventas'])
        print("\n")
        print("Ranking de productos:")
        listar_ranking(datos['ranking'])
        print("\n")
        print("Precio promedio por producto:")
        listar_precios(datos['precios'])

ventas = Ventas("datos.dat")
resumen = ventas.resumen_ventas(ventas.datos)
ventas.informe_ventas(resumen)
```

Escribe una funci√≥n en Python que encuentre los valores de `a`, `b`, y `c` para que la funci√≥n cuadr√°tica `f(x) = a x^2 + b x + c` pase exactamente por los siguientes puntos:

| x  | y  |
|---:|---:|
|  0 |  0 |
|  1 |  8 |
|  2 | 12 |
|  3 | 12 |
|  5 |  0 |

### Requisitos:
- La funci√≥n debe explorar posibles valores de `a`, `b`, y `c` utilizando un m√©todo de prueba y error.
- Debe devolver los valores que hagan que la diferencia entre la funci√≥n `f(x)` y los valores medidos `y` sea exactamente cero para cada punto.

> **Pista**: Los valores de `a`, `b`, y `c` son n√∫meros peque√±os.

La idea es implementar el mismo algoritmo que se uso en el TP1 pero usando NumPy en lugar de Python puro.

```python
import numpy as np
def f(x, coeficientes):
    a,b,c = coeficientes
    return a*x**2 + b*x + c

def error(y, y_pred):
    return y - y_pred

X = np.array([0,1,2,3,5])
Y = np.array([0,8,12,12,0])

def buscar_coeficientes():
    puntosFuncion = np.array([
        (0, 0), 
        (1, 8), 
        (2, 12), 
        (3, 12), 
        (5, 0)
    ])

    valores_a = np.arange(-10, 11)
    valores_b = np.arange(-10, 11)
    valores_c = np.arange(-10, 11)
    
    for a in valores_a:
        for b in valores_b:
            for c in valores_c:
                bandera = True
                
                for x, y in puntosFuncion:
                    if a * x**2 + b * x + c != y:
                        bandera = False
                        break
                        
                if bandera:
                    return int(a), int(b), int(c)

coeficientes = buscar_coeficientes()
coeficientes
```

### Ejercicio 9: Resolver el ejercicio 3 del TP1 usando NumPy
Buscar los coeficientes de la funci√≥n que minimice la suma de los cuadrados de las diferencias entre los valores medidos y los valores de la funci√≥n.

1. Crear un array con los coeficientes elegidos al azar (usar `randint(-10,10,3)`).
2. Calcular el valor de la funci√≥n y el error correspondiente.
3. Mientras que el error sea mayor a 1:
    1. Definir nuevos coeficientes agreg√°ndoles un peque√±o valor al azar a los coeficientes actuales (aprendizaje = 0.001).
    2. Si el error para los nuevos coeficientes es menor que el anterior, reemplazar los coeficientes actuales por los nuevos.


```python
import numpy as np
from numpy.random import randint

def f(x, coeficientes):
    a,b,c = coeficientes
    return a*x**2 + b*x + c

def error(y, y_pred):
    return np.sum((y - y_pred)**2)

X = np.array([0, 1, 2, 3, 5])
Y = np.array([0, 8,12,11, 1]) # Observar que no son los mismos valores que en el ejemplo anterior

def buscar_coeficientes():
    X = np.array([0, 1, 2, 3, 5])
    Y = np.array([0, 8, 12, 12, 0])
    
    e = np.inf
    coeficientes = np.random.randint(-10, 10, 3)
    aprendizaje = 0.001
    n = 0

    while e > 1:
        nuevos = coeficientes + aprendizaje * np.random.randint(-1, 2, 3)
        y = f(X, nuevos)
        ne = error(Y, y)
        
        if ne < e:
            e = ne
            coeficientes = nuevos
        
        n += 1
    
    return coeficientes

coeficientes = buscar_coeficientes()
coeficientes
```
