## Productos
- `codigo`: 4 digitos
- `nombre`: 1 a 100 caracteres
- `precio`: 10 a 10000
- `tipo`: 0 a 20 caracteres
- `cantidad`: 0 a 100 
- `valorTotal`: cantidad * precio

```python
class Producto:
    def __init__(self, codigo, nombre, precio, tipo, cantidad):
        # Validar codigo: debe ser un string de 4 dígitos
        if not (isinstance(codigo, str) and codigo.isdigit() and len(codigo) == 4):
            raise ValueError("El código debe ser un string de 4 dígitos.")
        
        # Validar nombre: debe tener entre 1 y 100 caracteres
        if not (isinstance(nombre, str) and 1 <= len(nombre) <= 100):
            raise ValueError("El nombre debe tener entre 1 y 100 caracteres.")
        
        # Validar precio: debe estar entre 10 y 10000
        if not (isinstance(precio, (int, float)) and 10 <= precio <= 10000):
            raise ValueError("El precio debe estar entre 10 y 10000.")
        
        # Validar tipo: debe tener entre 0 y 20 caracteres
        if not (isinstance(tipo, str) and len(tipo) <= 20):
            raise ValueError("El tipo debe tener entre 0 y 20 caracteres.")
        
        # Validar cantidad: debe estar entre 0 y 100
        if not (isinstance(cantidad, int) and 0 <= cantidad <= 100):
            raise ValueError("La cantidad debe estar entre 0 y 100.")
        
        # Asignar valores a los atributos privados
        self._codigo = codigo
        self._nombre = nombre
        self._precio = precio
        self._tipo = tipo
        self._cantidad = cantidad
    
    # Propiedad para obtener el código
    @property
    def codigo(self):
        return self._codigo
    
    # Propiedad para obtener el nombre
    @property
    def nombre(self):
        return self._nombre
    
    # Propiedad para obtener el precio
    @property
    def precio(self):
        return self._precio
    
    # Setter para el precio: rechaza valores negativos
    @precio.setter
    def precio(self, value):
        if value < 0:
            return
        self._precio = value
    
    # Propiedad para obtener el tipo
    @property
    def tipo(self):
        return self._tipo
    
    # Propiedad para obtener la cantidad
    @property
    def cantidad(self):
        return self._cantidad
    
    # Setter para la cantidad: rechaza valores negativos
    @cantidad.setter
    def cantidad(self, value):
        if value < 0:
            return
        self._cantidad = value
    
    # Método para calcular el valor total (cantidad * precio)
    def valorTotal(self):
        return self._cantidad * self._precio
```

# Ofertas
Debe permitir aplicar oferctas a codigos espeficicos de productos y a tipos de productos
- `descripcion`: 1 a 100 caracteres
- `codigos`: lista de codigos de productos
- `tipos`: lista de tipo de producto
- `esAplicable(producto, cantidad)`: retorna si la oferta es aplicable a ese producto
- `aplicar(producto, cantidad)`: retorna el precio final del producto con la oferta aplicada


```python
class Oferta:
    def __init__(self, descripcion, codigos=[], tipos=[]):
        # Validar descripcion: debe tener entre 1 y 100 caracteres
        if not (isinstance(descripcion, str) and 1 <= len(descripcion) <= 100):
            raise ValueError("La descripción debe tener entre 1 y 100 caracteres.")
        
        # Validar codigos: debe ser una lista
        if not isinstance(codigos, list):
            raise ValueError("Los códigos deben ser una lista.")
        
        # Validar tipos: debe ser una lista
        if not isinstance(tipos, list):
            raise ValueError("Los tipos deben ser una lista.")
        
        self.descripcion = descripcion
        self.codigos = codigos
        self.tipos = tipos
    
    def esAplicable(self, producto, cantidad):
        # Verificar si el producto es aplicable a la oferta
        return producto.codigo in self.codigos or producto.tipo in self.tipos
    
    def aplicar(self, producto, cantidad):
        # Método abstracto para aplicar la oferta
        raise NotImplementedError("Este método debe ser implementado por las subclases")
    
    def calcularDescuento(self, producto, cantidad):
        # Método abstracto para calcular el descuento
        raise NotImplementedError("Este método debe ser implementado por las subclases")

class OfertaDescuento(Oferta):
    def __init__(self, descuento, descripcion="Descuento", codigos=[], tipos=[]):
        super().__init__(descripcion, codigos, tipos)
        # Validar descuento: debe ser un número entre 0 y 100
        if not (isinstance(descuento, (int, float)) and 0 <= descuento <= 100):
            raise ValueError("El descuento debe estar entre 0 y 100.")
        
        self.descuento = descuento
    
    def aplicar(self, producto, cantidad):
        # Aplicar el descuento al precio del producto
        if self.esAplicable(producto, cantidad):
            return producto.precio * cantidad * (1 - self.descuento / 100)
        return producto.precio * cantidad
    
    def calcularDescuento(self, producto, cantidad):
        # Calcular el descuento aplicable
        if self.esAplicable(producto, cantidad):
            return producto.precio * cantidad * (self.descuento / 100)
        return 0

class Oferta2x1(Oferta):
    def __init__(self, descripcion="2x1", codigos=[], tipos=[]):
        super().__init__(descripcion, codigos, tipos)
    
    def aplicar(self, producto, cantidad):
        # Aplicar la oferta 2x1 al precio del producto
        if self.esAplicable(producto, cantidad):
            return producto.precio * (cantidad // 2 + cantidad % 2)
        return producto.precio * cantidad
    
    def calcularDescuento(self, producto, cantidad):
        # Calcular el descuento aplicable
        if self.esAplicable(producto, cantidad):
            return producto.precio * (cantidad // 2)
        return 0
```

# Catalogo
- `leer(archivo) `    : Carga los productos desde el archivo
- `guardar(archivo)`  : Guarda los productos en el archivo
- `agregar(producto)` : Agrega un producto al catalogo
- `buscar(codigo)`    : Busca un producto por codigo o None si no existe
- `registrarOferta(oferta)`  : Registra una oferta
- `buscarOferta(producto, cantidad)`: Busca una oferta por codigo o None si no existe
- `calcularDescuento(producto, cantidad)`: Calcula el descuento de una oferta
- `cantidadProductos`: Retorna la cantidad de productos en el catalogo
- `cantidadUnidades`: Retorna la cantidad de unidades en el catalogo
- `valorTotal`: retorna el valor total del catalogo sin descuentos
- `informe()`: retorna un string con el informe del catalogo 

```python
import csv  # Importa el módulo csv para trabajar con archivos CSV

# Clase base para las ofertas
class Oferta:
    def __init__(self, descripcion, codigos=[], tipos=[]):
        # Inicializa la oferta con una descripción, una lista de códigos y una lista de tipos
        if not (isinstance(descripcion, str) and 1 <= len(descripcion) <= 100):
            raise ValueError("La descripción debe tener entre 1 y 100 caracteres.")
        if not isinstance(codigos, list):
            raise ValueError("Los códigos deben ser una lista.")
        if not isinstance(tipos, list):
            raise ValueError("Los tipos deben ser una lista.")
        
        self.descripcion = descripcion  # Descripción de la oferta
        self.codigos = codigos  # Lista de códigos de productos aplicables
        self.tipos = tipos  # Lista de tipos de productos aplicables
    
    def esAplicable(self, producto, cantidad):
        # Verifica si la oferta es aplicable a un producto dado
        return producto.codigo in self.codigos or producto.tipo in self.tipos
    
    def aplicar(self, producto, cantidad):
        # Método abstracto para aplicar la oferta (debe ser implementado por subclases)
        raise NotImplementedError("Este método debe ser implementado por las subclases")
    
    def calcularDescuento(self, producto, cantidad):
        # Método abstracto para calcular el descuento (debe ser implementado por subclases)
        raise NotImplementedError("Este método debe ser implementado por las subclases")

# Clase para la oferta 2x1
class Oferta2x1(Oferta):
    def __init__(self, descripcion="Oferta 2x1", codigos=[], tipos=[]):
        # Inicializa la oferta 2x1 con una descripción, una lista de códigos y una lista de tipos
        super().__init__(descripcion, codigos, tipos)
    
    def aplicar(self, producto, cantidad):
        # Aplica la oferta 2x1 al precio del producto
        if self.esAplicable(producto, cantidad):
            return producto.precio * (cantidad // 2 + cantidad % 2)
        return producto.precio * cantidad
    
    def calcularDescuento(self, producto, cantidad):
        # Calcula el descuento aplicable para la oferta 2x1
        if self.esAplicable(producto, cantidad):
            return producto.precio * (cantidad // 2)
        return 0

# Clase para el catálogo de productos
class Catalogo:
    def __init__(self):
        self.productos = []  # Lista de productos en el catálogo
        self.ofertas = []  # Lista de ofertas registradas
    
    def agregar(self, producto):
        # Agrega un producto al catálogo
        self.productos.append(producto)
    
    def buscar(self, codigo):
        # Busca un producto por su código
        for producto in self.productos:
            if producto.codigo == codigo:
                return producto
        return None
    
    def registrarOferta(self, oferta):
        # Registra una oferta en el catálogo
        self.ofertas.append(oferta)
    
    def buscarOferta(self, producto, cantidad):
        # Busca una oferta aplicable a un producto dado
        for oferta in self.ofertas:
            if oferta.esAplicable(producto, cantidad):
                return oferta
        return None
    
    def calcularDescuento(self, producto, cantidad):
        # Calcula el descuento aplicable a un producto dado
        oferta = self.buscarOferta(producto, cantidad)
        if oferta:
            return oferta.calcularDescuento(producto, cantidad)
        return 0
    
    @property
    def cantidadProductos(self):
        # Retorna la cantidad de productos en el catálogo
        return len(self.productos)
    
    @property
    def cantidadUnidades(self):
        # Retorna la cantidad total de unidades en el catálogo
        return sum(producto.cantidad for producto in self.productos)
    
    @property
    def valorTotal(self):
        # Retorna el valor total del catálogo sin descuentos
        return sum(producto.valorTotal() for producto in self.productos)
    
    def informe(self):
        # Genera un informe del catálogo
        tipos = set(producto.tipo for producto in self.productos)
        informe = (
            f"Cantidad de productos: {self.cantidadProductos}\n"
            f"Cantidad de unidades: {self.cantidadUnidades}\n"
            f"Precio Promedio: {self.valorTotal / self.cantidadUnidades:.2f}\n"
            f"Valor total: {self.valorTotal:.2f}\n"
            f"Tipos de productos: {', '.join(tipos)}\n"
            f"Ofertas:\n"
        )
        for oferta in self.ofertas:
            informe += f"  - {oferta.descripcion}\n"
        return informe
    
    def guardar(self, archivo):
        # Guarda los productos del catálogo en un archivo CSV
        with open(archivo, 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(['codigo', 'nombre', 'precio', 'tipo', 'cantidad'])
            for producto in self.productos:
                writer.writerow([producto.codigo, producto.nombre, producto.precio, producto.tipo, producto.cantidad])
    
    @classmethod
    def leer(cls, archivo):
        # Lee los productos desde un archivo CSV y los carga en un nuevo catálogo
        catalogo = cls()
        with open(archivo, 'r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                producto = Producto(row['codigo'], row['nombre'], float(row['precio']), row['tipo'], int(row['cantidad']))
                catalogo.agregar(producto)
        return catalogo
    
    def vender(self, producto, cantidad):
        # Vende una cantidad de unidades de un producto
        if producto in self.productos and producto.cantidad >= cantidad:
            producto.cantidad -= cantidad
        else:
            raise ValueError("Cantidad insuficiente para vender")
```

```python
## NO MODIFIQUE ESTE CODIGO ##

# Prueba del catálogo 

catalogo = Catalogo()
p1 = Producto('0001', 'Coca Cola',  1500, 'gaseosa', 10)
p2 = Producto('0002', 'Pepsi Cola', 1200, 'gaseosa', 20)
p3 = Producto('0003', 'Sonrisa',    1200, 'galleta', 30)
p4 = Producto('0004', 'Oreo',       2300, 'galleta', 40)

## Agregar productos al catalogo 
catalogo.agregar(p1)
catalogo.agregar(p2)
catalogo.agregar(p3)
catalogo.agregar(p4)

assert catalogo.cantidadProductos == 4
assert catalogo.cantidadUnidades == 100

assert catalogo.valorTotal == 167000

## Calcular descuentos segun las ofertas registradas
assert catalogo.calcularDescuento(p1, 5) == 0
assert catalogo.calcularDescuento(p2, 5) == 0

# Ofertas no acumulables 
catalogo.registrarOferta(Oferta2x1(tipos=['galleta']))
catalogo.registrarOferta(OfertaDescuento(10, codigos=['0001', '0003']))

assert catalogo.calcularDescuento(p1, 5) == 750
assert catalogo.calcularDescuento(p2, 5) == 0
assert catalogo.calcularDescuento(p3, 5) == 2400

assert catalogo.valorTotal == 167000.0
catalogo.guardar('catalogo-prueba.csv') ## Guardar datos antes de vender

# Vender afecta la cantidad de unidades y el valor total
catalogo.vender(p3, 3)   

# Verificar que el informe se genere correctamente

informe = catalogo.informe()
assert "Cantidad de productos: " in informe
assert "Cantidad de unidades: " in informe
assert "Precio Promedio: " in informe
assert "Valor total: " in informe
assert "Tipos de productos: " in informe
assert "gaseosa" in informe
assert "galleta" in informe
assert "Ofertas:" in informe 
assert "Oferta 2x1" in informe
assert catalogo.cantidadUnidades == 97
assert catalogo.valorTotal == 163400

# Buscar por código
assert catalogo.buscar('0001') == p1
assert catalogo.buscar('0002') == p2
assert catalogo.buscar('0099') is None 

# Recuperar los datos guardados  
c2 = Catalogo.leer('catalogo-prueba.csv')

assert c2.cantidadProductos == 4
assert c2.cantidadUnidades == 100

# Valor antes de guardar
assert c2.valorTotal == 167000.0

print("Prueba pasada exitosamente!")
```

# Cliente
- `nombre`: 1 a 100 caracteres
- `cuit`: 13 digitos (formato XX-XXXXXXXX-X)

```python
class Cliente:
    def __init__(self, nombre, cuit):
        self._nombre = None  # Inicializa el atributo privado _nombre
        self._cuit = None    # Inicializa el atributo privado _cuit
        self.nombre = nombre  # Utiliza el setter para validar y asignar el nombre
        self.cuit = cuit      # Utiliza el setter para validar y asignar el cuit
    
    @property
    def nombre(self):
        # Getter para el atributo nombre
        return self._nombre
    
    @nombre.setter
    def nombre(self, value):
        # Setter para el atributo nombre con validación
        if isinstance(value, str) and 1 <= len(value) <= 100:
            self._nombre = value  # Asigna el valor si es válido
    
    @property
    def cuit(self):
        # Getter para el atributo cuit
        return self._cuit
    
    @cuit.setter
    def cuit(self, value):
        # Setter para el atributo cuit con validación
        if isinstance(value, str) and len(value) == 13 and value[2] == '-' and value[11] == '-' and value[:2].isdigit() and value[3:11].isdigit() and value[12].isdigit():
            self._cuit = value  # Asigna el valor si es válido
```

```python
import csv

class Catalogo:
    def __init__(self):
        self.productos = []
        self.ofertas = []
    
    
    def agregar(self, *productos):
        # Método para agregar múltiples productos al catálogo
        for producto in productos:
            self.productos.append(producto)
    
    def buscar(self, codigo):
        for producto in self.productos:
            if producto.codigo == codigo:
                return producto
        return None
    
    def registrarOferta(self, oferta):
        self.ofertas.append(oferta)
    
    def buscarOferta(self, producto, cantidad):
        for oferta in self.ofertas:
            if oferta.esAplicable(producto, cantidad):
                return oferta
        return None
    
    def calcularDescuento(self, producto, cantidad):
        oferta = self.buscarOferta(producto, cantidad)
        if oferta:
            return oferta.calcularDescuento(producto, cantidad)
        return 0
    
    @property
    def cantidadProductos(self):
        return len(self.productos)
    
    @property
    def cantidadUnidades(self):
        return sum(producto.cantidad for producto in self.productos)
    
    @property
    def valorTotal(self):
        return sum(producto.valorTotal() for producto in self.productos)
    
    def informe(self):
        tipos = set(producto.tipo for producto in self.productos)
        informe = (
            f"Cantidad de productos: {self.cantidadProductos}\n"
            f"Cantidad de unidades: {self.cantidadUnidades}\n"
            f"Precio Promedio: {self.valorTotal / self.cantidadUnidades:.2f}\n"
            f"Valor total: {self.valorTotal:.2f}\n"
            f"Tipos de productos: {', '.join(tipos)}\n"
            f"Ofertas:\n"
        )
        for oferta in self.ofertas:
            informe += f"  - {oferta.descripcion}\n"
        return informe
    
    def guardar(self, archivo):
        with open(archivo, 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(['codigo', 'nombre', 'precio', 'tipo', 'cantidad'])
            for producto in self.productos:
                writer.writerow([producto.codigo, producto.nombre, producto.precio, producto.tipo, producto.cantidad])
    
    @classmethod
    def leer(cls, archivo):
        catalogo = cls()
        with open(archivo, 'r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                producto = Producto(row['codigo'], row['nombre'], float(row['precio']), row['tipo'], int(row['cantidad']))
                catalogo.agregar(producto)
        return catalogo
    
    def vender(self, producto, cantidad):
        if producto in self.productos and producto.cantidad >= cantidad:
            producto.cantidad -= cantidad
        else:
            raise ValueError("Cantidad insuficiente para vender")
    

class Factura:
    _ultimo_numero = 100  # Variable de clase para llevar el control del último número de factura

    @classmethod
    def ultimaFactura(cls, numero):
        # Método de clase para establecer el último número de factura
        cls._ultimo_numero = numero

    @classmethod
    def nuevoNumero(cls):
        # Método de clase para obtener un nuevo número de factura
        cls._ultimo_numero += 1
        return cls._ultimo_numero

    def __init__(self, catalogo, cliente):
        self.numero = Factura.nuevoNumero()  # Asigna un nuevo número de factura
        self.catalogo = catalogo  # Asigna el catálogo de productos
        self.cliente = cliente  # Asigna el cliente
        self.items = {}  # Diccionario para almacenar los productos y sus cantidades

    def agregar(self, producto, cantidad):
        # Método para agregar productos a la factura
        if producto in self.items:
            self.items[producto] += cantidad  # Si el producto ya está en la factura, incrementa la cantidad
        else:
            self.items[producto] = cantidad  # Si el producto no está en la factura, lo agrega con la cantidad inicial
        self.catalogo.vender(producto, cantidad)  # Descontar las unidades vendidas del catálogo

    @property
    def cantidadProductos(self):
        # Retorna la cantidad de productos diferentes en la factura
        return len(self.items)

    @property
    def cantidadUnidades(self):
        # Retorna la cantidad total de unidades en la factura
        return sum(self.items.values())

    @property
    def subtotal(self):
        # Calcula el subtotal de la factura sin descuentos
        return sum(producto.precio * cantidad for producto, cantidad in self.items.items())

    @property
    def descuentos(self):
        # Calcula el total de descuentos aplicables a la factura
        return sum(self.catalogo.calcularDescuento(producto, cantidad) for producto, cantidad in self.items.items())

    @property
    def total(self):
        # Calcula el total de la factura después de aplicar los descuentos
        return self.subtotal - self.descuentos

    def imprimir(self):
        # Genera una representación en texto de la factura
        impresion = f"Factura N°: {self.numero}\nCliente: {self.cliente.nombre}\nCUIT: {self.cliente.cuit}\n\n"
        impresion += "Productos:\n"
        for producto, cantidad in self.items.items():
            impresion += f"{cantidad} {producto.nombre} - ${producto.precio * cantidad:.2f}\n"
        impresion += f"\nSubtotal: ${self.subtotal:.2f}\n"
        impresion += f"Descuentos: ${self.descuentos:.2f}\n"
        impresion += f"TOTAL: ${self.total:.2f}\n"
        
        # Agregar las descripciones de las ofertas aplicadas
        impresion += "\nOfertas aplicadas:\n"
        for producto, cantidad in self.items.items():
            oferta = self.catalogo.buscarOferta(producto, cantidad)
            if oferta:
                if isinstance(oferta, OfertaDescuento):
                    impresion += f"Descuento {oferta.descuento}%\n"
                else:
                    impresion += f"{oferta.descripcion}\n"
        
        return impresion

class Producto:
    def __init__(self, codigo, nombre, precio, tipo, cantidad):
        if not (isinstance(codigo, str) and codigo.isdigit() and len(codigo) == 4):
            raise ValueError("El código debe ser un string de 4 dígitos.")
        if not (isinstance(nombre, str) and 1 <= len(nombre) <= 100):
            raise ValueError("El nombre debe tener entre 1 y 100 caracteres.")
        if not (isinstance(precio, (int, float)) and 10 <= precio <= 10000):
            raise ValueError("El precio debe estar entre 10 y 10000.")
        if not (isinstance(tipo, str) and len(tipo) <= 20):
            raise ValueError("El tipo debe tener entre 0 y 20 caracteres.")
        if not (isinstance(cantidad, int) and 0 <= cantidad <= 100):
            raise ValueError("La cantidad debe estar entre 0 y 100.")
        
        self._codigo = codigo
        self._nombre = nombre
        self._precio = precio
        self._tipo = tipo
        self._cantidad = cantidad
    
    @property
    def codigo(self):
        return self._codigo
    
    @property
    def nombre(self):
        return self._nombre
    
    @property
    def precio(self):
        return self._precio
    
    @precio.setter
    def precio(self, value):
        if value < 0:
            return
        self._precio = value
    
    @property
    def tipo(self):
        return self._tipo
    
    @property
    def cantidad(self):
        return self._cantidad
    
    @cantidad.setter
    def cantidad(self, value):
        if value < 0:
            return
        self._cantidad = value
    
    def valorTotal(self):
        return self._cantidad * self._precio

class Oferta:
    def __init__(self, descripcion, codigos=[], tipos=[]):
        if not (isinstance(descripcion, str) and 1 <= len(descripcion) <= 100):
            raise ValueError("La descripción debe tener entre 1 y 100 caracteres.")
        if not isinstance(codigos, list):
            raise ValueError("Los códigos deben ser una lista.")
        if not isinstance(tipos, list):
            raise ValueError("Los tipos deben ser una lista.")
        
        self.descripcion = descripcion
        self.codigos = codigos
        self.tipos = tipos
    
    def esAplicable(self, producto, cantidad):
        return producto.codigo in self.codigos or producto.tipo in self.tipos
    
    def aplicar(self, producto, cantidad):
        raise NotImplementedError("Este método debe ser implementado por las subclases")
    
    def calcularDescuento(self, producto, cantidad):
        raise NotImplementedError("Este método debe ser implementado por las subclases")

class OfertaDescuento(Oferta):
    def __init__(self, descuento, descripcion="Descuento", codigos=[], tipos=[]):
        super().__init__(descripcion, codigos, tipos)
        if not (isinstance(descuento, (int, float)) and 0 <= descuento <= 100):
            raise ValueError("El descuento debe estar entre 0 y 100.")
        
        self.descuento = descuento
    
    def aplicar(self, producto, cantidad):
        if self.esAplicable(producto, cantidad):
            return producto.precio * cantidad * (1 - self.descuento / 100)
        return producto.precio * cantidad
    
    def calcularDescuento(self, producto, cantidad):
        if self.esAplicable(producto, cantidad):
            return producto.precio * cantidad * (self.descuento / 100)
        return 0

class Oferta2x1(Oferta):
    def __init__(self, descripcion="Oferta 2x1", codigos=[], tipos=[]):
        super().__init__(descripcion, codigos, tipos)
    
    def aplicar(self, producto, cantidad):
        if self.esAplicable(producto, cantidad):
            return producto.precio * (cantidad // 2 + cantidad % 2)
        return producto.precio * cantidad
    
    def calcularDescuento(self, producto, cantidad):
        if self.esAplicable(producto, cantidad):
            return producto.precio * (cantidad // 2)
        return 0

class Cliente:
    def __init__(self, nombre, cuit):
        self._nombre = None
        self._cuit = None
        self.nombre = nombre  # Utiliza el setter para validar
        self.cuit = cuit      # Utiliza el setter para validar
    
    @property
    def nombre(self):
        return self._nombre
    
    @nombre.setter
    def nombre(self, value):
        if isinstance(value, str) and 1 <= len(value) <= 100:
            self._nombre = value
    
    @property
    def cuit(self):
        return self._cuit
    
    @cuit.setter
    def cuit(self, value):
        if isinstance(value, str) and len(value) == 13 and value[2] == '-' and value[11] == '-' and value[:2].isdigit() and value[3:11].isdigit() and value[12].isdigit():
            self._cuit = value
```
