Las **cartas** deben registra el numero, palo y valor de la carta. 
El valor debe reflejar la jerarquía de las cartas en el juego del truco.

**Mazo**: Representan las cartas españolas de 1 al 12, en los palos de oro, copa, espada y basto.

Debe existir una función que permita barajar las cartas.
y otra función que permita repartir las cartas a los jugadores.

```python
import random

jerarquia = {
    (1, "espada"): 1,
    (1, "basto"): 2,
    (7, "espada"): 3,
    (7, "oro"): 4,
    (3, "espada"): 5,
    (3, "basto"): 5, 
    (3, "oro"): 5, 
    (3, "copa"): 5,
    (2, "espada"): 6, 
    (2, "basto"): 6, 
    (2, "oro"): 6, 
    (2, "copa"): 6,
    (1, "oro"): 7, 
    (1, "copa"): 7,
    (12, "espada"): 8, 
    (12, "basto"): 8, 
    (12, "oro"): 8, 
    (12, "copa"): 8,
    (11, "espada"): 9, 
    (11, "basto"): 9, 
    (11, "oro"): 9, 
    (11, "copa"): 9,
    (10, "espada"): 10, 
    (10, "basto"): 10, 
    (10, "oro"): 10, 
    (10, "copa"): 10,
    (7, "copa"): 11, 
    (7, "basto"): 11,
    (6, "espada"): 12, 
    (6, "basto"): 12, 
    (6, "oro"): 12, 
    (6, "copa"): 12,
    (5, "espada"): 13, 
    (5, "basto"): 13, 
    (5, "oro"): 13, 
    (5, "copa"): 13,
    (4, "espada"): 14, 
    (4, "basto"): 14, 
    (4, "oro"): 14, 
    (4, "copa"): 14
}

class Mazo:
    def __init__(self, cartas):
        self.cartas = list(cartas.keys())  

    def mezclar(self):
        
        self.cartas = random.sample(self.cartas, len(self.cartas))

    def repartir(self, cantidad):
        repartidas = self.cartas[:cantidad]
        self.cartas = self.cartas[cantidad:]  
        return repartidas

m = Mazo(jerarquia)

m.mezclar()
print(m.repartir(3))
```

**Jugador**: Deben tener un nombre y las cartas que tiene en la mano. Debe acumular los puntos ganados.

Nota: Para simplificar el problema, el jugador elige la carta a jugar de forma aleatoria.

```python


class Jugador:
    def __init__(self, nombre):
        self.nombre = nombre
        self.cartas = []

    def recibir_cartas(self, cartas):
        for carta in cartas:
            self.cartas.append(carta)

    def jugar_carta(self):
        if self.cartas:  
            return self.cartas.pop(0)
        else:
            return None  

m = Mazo(jerarquia)
m.mezclar()
j = Jugador("Juan")
j.recibir_cartas(m.repartir(3))

carta_jugada = j.jugar_carta()
print(f"{j.nombre} jugó la carta: {carta_jugada}")
```

**Partida**: Debe tener dos jugadores, un mazo de cartas y un puntaje. Debe permitir jugar una carta por ronda y definir el ganador de la ronda. Debe permitir jugar hasta que se terminen las cartas del mazo.

```python
import random

# Jerarquía de las cartas
jerarquia = {
    (1, "espada"): 1,
    (1, "basto"): 2,
    (7, "espada"): 3,
    (7, "oro"): 4,
    (3, "espada"): 5,
    (3, "basto"): 5,
    (3, "oro"): 5,
    (3, "copa"): 5,
    (2, "espada"): 6,
    (2, "basto"): 6,
    (2, "oro"): 6,
    (2, "copa"): 6,
    (1, "oro"): 7,
    (1, "copa"): 7,
    (12, "espada"): 8,
    (12, "basto"): 8,
    (12, "oro"): 8,
    (12, "copa"): 8,
    (11, "espada"): 9,
    (11, "basto"): 9,
    (11, "oro"): 9,
    (11, "copa"): 9,
    (10, "espada"): 10,
    (10, "basto"): 10,
    (10, "oro"): 10,
    (10, "copa"): 10,
    (7, "copa"): 11,
    (7, "basto"): 11,
    (6, "espada"): 12,
    (6, "basto"): 12,
    (6, "oro"): 12,
    (6, "copa"): 12,
    (5, "espada"): 13,
    (5, "basto"): 13,
    (5, "oro"): 13,
    (5, "copa"): 13,
    (4, "espada"): 14,
    (4, "basto"): 14,
    (4, "oro"): 14,
    (4, "copa"): 14
}

class Mazo:
    def __init__(self, cartas):
        self.cartas = cartas
        
    def mezclar(self):
        random.shuffle(self.cartas)

    def repartir(self, cantidad):
        return [self.cartas.pop() for _ in range(min(cantidad, len(self.cartas)))]

class Jugador:
    def __init__(self, nombre):
        self.nombre = nombre
        self.cartas = []
        self.puntos = 0

    def recibir_cartas(self, cartas):
        for carta in cartas:
            self.cartas.append(carta)

    def jugar_carta(self):
        if self.cartas:
            return self.cartas.pop(0) 
        return None

    def sumar_puntos(self, puntos):
        self.puntos += puntos

class Partida:
    def __init__(self, jugador1, jugador2):
        self.jugador1 = jugador1
        self.jugador2 = jugador2
        self.mazo = Mazo(list(jerarquia.keys()))  
        self.mazo.mezclar()  

    def jugar_ronda(self):
        cartas_jugador1 = self.mazo.repartir(3)
        cartas_jugador2 = self.mazo.repartir(3)

        self.jugador1.recibir_cartas(cartas_jugador1)
        self.jugador2.recibir_cartas(cartas_jugador2)

        carta1 = self.jugador1.jugar_carta()
        carta2 = self.jugador2.jugar_carta()

        if carta1 and carta2:
            valor1 = jerarquia[carta1]
            valor2 = jerarquia[carta2]
            if valor1 > valor2:
                self.jugador1.sumar_puntos(1)
                return f"{self.jugador1.nombre} gana la ronda con {carta1}!"
            elif valor1 < valor2:
                self.jugador2.sumar_puntos(1)
                return f"{self.jugador2.nombre} gana la ronda con {carta2}!"
            else:
                return "La ronda es un empate."
        return "No hay cartas suficientes para jugar."

    def jugar(self):
        while self.jugador1.puntos < 15 and self.jugador2.puntos < 15:
            resultado = self.jugar_ronda()
            print(resultado)

        if self.jugador1.puntos >= 15:
            print(f"{self.jugador1.nombre} gana la partida con {self.jugador1.puntos} puntos!")
        else:
            print(f"{self.jugador2.nombre} gana la partida con {self.jugador2.puntos} puntos!")

p = Partida(Jugador("Juan"), Jugador("Pedro"))
p.jugar()
```
